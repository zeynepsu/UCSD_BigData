## Clustering using KD-trees

Here is a visualization of the data using a **4-dimensional KD tree**.
The 4 dimensions are :
1. Longitude value
2. Latitude value
3. A label 1-30. Each label corresponds to an equal width **vertical strip** on the world map.
4. A label 1-30. Each label corresponds to an equal width **horizontal strip** on the world map.

The motivation behind feature number 3 is that it will capture stations with similar weather patterns.
The motivation behind feature 4 was that it will capture areas of the map that occur in the same country.

```python
lon=station_info.ix[:,'longitude'].values
lat=station_info.ix[:,'latitude'].values 
type(lat)
```

```python
def leaves(tree):
    if hasattr(tree, 'idx'):#tree is a leafnode
        return [tree.idx]
    else:
        return leaves(tree.less) + leaves(tree.greater)
```

```python
from scipy import spatial
data = np.zeros((len(lat),4))
data[:,0] = lat
data[:,1] = lon
```

```python
lat_bins = np.linspace(-90,90,30)
lon_bins = np.linspace(-180,180,30)
digitized_lats = np.digitize(lat,lat_bins)
digitized_lons = np.digitize(lon,lon_bins)

data[:,2] = digitized_lats
data[:,3] = digitized_lons
```

```python
import sys
sys.setrecursionlimit(100000)
kdt = spatial.KDTree(data, leafsize=1000)
```

```python
color = [0]*len(lat)
c = 0
for i in leaves(kdt.tree):
    for j in i:
        color[j] = c   
    c = c+ 1
 
print len(leaves(kdt.tree))
```
