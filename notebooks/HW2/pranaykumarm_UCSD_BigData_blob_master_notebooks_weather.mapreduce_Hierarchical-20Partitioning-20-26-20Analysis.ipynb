# Hierarchical partitioning using kd-tree

I wrote a function that performs hierarchical partitioning based on the number of measurements available at every station. The stations are partitioned into buckets such that every bucket has the total number of measurements less than a threshold value. 

At each step, the data is partitioned in such a way that the difference between the two partitions is less than 20% of the minimum number of measurements over the two partitions.

Partitioning is done along latitudes or longitudes aletrnatively. The user has a choice to choose along which axis the partiotioning starts

Additionally, the tree structure is preserved (in the form of nested loops) and siblings are closer to each other. This will be useful when we try to merge two adjacent regions (if necessary).

Execution time = 2sec for all the given 85283 stations!

```python
bucket_no = {}
bucket_val = 1

def kdtree(level, ax, data):
    global bucket_no, bucket_val
    print 'level', level
    delta = 1.0
    buckets = []
    thresh = 600000
    k = data.shape[1]-1
    maxax = max(data[:,ax])
    minax = min(data[:,ax])
    print 'min coordinate', minax, 'max coordinate', maxax
    midax = (maxax+minax)/2
    selecthi = data[:,ax]>=midax
    selectlo = data[:,ax]<midax
    recordshi = sum(data[:,-2][selecthi])
    recordslo = sum(data[:,-2][selectlo])
    
    lastax = midax
    curax = midax
    trial = 0
    print recordshi, recordslo, recordshi+recordslo
    while abs(recordshi-recordslo)/min(recordshi,recordslo) > 0.2:
        if recordshi>recordslo:
            newax = curax+delta
        else:
            newax = curax-delta
        if newax==lastax:
            delta = delta/2
        lastax = curax
        curax = newax
        selecthi = data[:,ax]>=curax
        selectlo = data[:,ax]<curax
        recordshi = sum(data[:,-2][selecthi])
        recordslo = sum(data[:,-2][selectlo])
        
        trial+=1
        if trial%100==0:
            print '100 more trials', lastax, newax
        if delta<0.0001:
            break
    print recordshi, recordslo, recordshi+recordslo
    ax = 1^ax
    level += 1
    if recordshi<=thresh:
        buckets.append(data[selecthi])
        for info in data[selecthi]:
            bucket_no[info[3]] = bucket_val
        bucket_val+=1
    else:
        buckets.append(kdtree(level, ax, data[selecthi]))
    if recordslo<=thresh:
        buckets.append(data[selectlo])
        for info in data[selectlo]:
            bucket_no[info[3]] = bucket_val
        bucket_val+=1
    else:
        buckets.append(kdtree(level, ax, data[selectlo]))
    
    return buckets
```

# PCA on individual leaves

The idea is to first concatenate the TMAX and TMIN values (creating a data vector of 730 features). When there are less than 365 measurements in a year, we can either interpolate the measurements or discard that data point. Each station coupled with each year gives a new data point.

The mapper identifies all the lines corresponding to TMAX and TMIN. These lines are concatenated by the reducer using (station,year) as the key. These lines are then categorized based on the buckets to which the stations belong to.

In the next step, using the bucket number as key, the reducer computes the mean and covariance matrix of each bucket (using the 730 feature vectors as data points) on the EMR. Then, I compute the eigenvectors of this matrix.

```python
%%writefile mr_pca_ind.py
#!/usr/bin/python
import sys
import numpy as np
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr

class mr_pca_ind(MRJob):
    def configure_options(self):
        super(mr_pca_ind,self).configure_options()
        self.add_file_option('--buckets')

    def reducer_init(self):
        f = gzip.open(self.options.buckets, "rb")
        pickleFile = pickle.Unpickler(f)
        self.buckets = pickleFile.load()
        f.close()
    
    def mapper_concatenate(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper-all',1)
            elements=line.split(',')
            number_defined=sum([e!='' for e in elements[3:]])
            if elements[0] != 'station':
                if number_defined == 365 and elements[1]=='TMAX':
                        yield ([elements[0],elements[2]],[1]+elements[:])
                if number_defined == 365 and elements[1]=='TMIN':
                        yield ([elements[0],elements[2]],[1]+elements[:])

        except Exception, e:
            stderr.write('Error in line:\n'+line)
            stderr.write(e)
            self.increment_counter('MrJob Counters','mapper-error',1)            

    def reducer_concatenate(self, station_year, counts):
        self.increment_counter('MrJob Counters','reducer',1)
        new=[]
        for v in counts:
            if v[2]=='TMAX':
                new=v[4:]+new
            else:
                new=new+v[4:]
        tmp='1'
        if len(new)==730:
            for s in new:
                tmp=tmp+' '+s
            tmp=tmp+' 1'
            yield (self.buckets[station_year[0]],[tmp])
            
    def steps(self):
        return [
            self.mr(mapper=self.mapper_concatenate,
                    reducer=self.reducer_concatenate)#,
            #self.mr(reducer=self.reducer_find_max_word)
        ]    

if __name__ == '__main__':
    mr_pca_ind.run()
```

# Merging using MDL criterion

Consider each neighboring pairs of regions and compute:

The eigenvectors for the two regions taken together, this will give you another set of eigen vectors v_1,\ldots,v_k that explaines 99% of the variance.

Suppose that the first region contains n1 measurements and requires k1 vectors to reach 99% of the variance. Similarly n2,k2 for the second region and n3 = n1 + n2,k3 for the combined regions.

The MDL criterion for when to merge the two regions is: $n_1\cdot k_1+(k_1+1)\cdot(2\times 365)+n_2\cdot k_2+(k_2+1)\cdot(2\times 365) > n_3\cdot k_3+(k_3+1)\cdot(2\times 365)$

If the condition holds, merge the two regions, otherwise, leave them separated.

Repeat until no more pairs of regions can be merged.
