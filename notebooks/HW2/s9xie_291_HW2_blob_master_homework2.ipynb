How to count measurements by Map-reduce?

Mapper: 
Key: station index ID, 
value: total number of measurements

Emit a key-value pair for "TMAX" and "TMIN". 

Reducer: Aggregate all measurement counts.

Combiner:
sums all the measurement counts before sending the data to the reducers when we got multiple lines with same station
