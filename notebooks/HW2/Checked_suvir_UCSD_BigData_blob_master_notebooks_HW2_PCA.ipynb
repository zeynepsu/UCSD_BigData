## Calculating covariance matrices using MRJOB

The code below was run using '-r local' on ALL.head.csv.

Due to various issues, it has not been run successfully on the entire data set.

**However, the heirarchical partitioning used was computed using entire data set.**

```python
%%writefile play_mrjob.py
"""
Doing PCA using MapReduce
"""
from mrjob.job import MRJob
import pandas as pd
import numpy as np
from sys import stderr
import re,pickle,base64,zlib
import cPickle,gzip
import random

logfile=stderr

def loads(eVal):
    """ Decode a string into a value """
    return pickle.loads(zlib.decompress(base64.b64decode(eVal)))

def dumps(Value):
    """ Encode a value as a string """
    return base64.b64encode(zlib.compress(pickle.dumps(Value),9))

class PCA(MRJob):
    def configure_options(self):
        super(PCA,self).configure_options()
        self.add_file_option('--stations')
        
    def get_buckets(self):
        f = gzip.open( self.options.stations, "rb" )
        pickleFile = cPickle.Unpickler(f)
        self.buckets = pickleFile.load()
        f.close()
        
    def mapper_1(self,_,value):
        vector = value.split(',')
        logfile.write('MAPPER:'+str(vector[:3])+'\t'+'\n')
        yield ' '.join(vector[:3]),vector[3:]
        
    def reducer_1(self,station,values):
        """
        Key for next step = station+year
        """
        s,m,y = station.split(' ')
        key_next = s+' '+y
        
        """
        1.Prepare Pandas Series from data, 
        2.Do interpolation, 
        3.Replace Nan with 0
        """
        #vector = np.matrix([0 if len(i)<1 else int(i) for i in [i for i in values][0]])
        vector = pd.Series([np.nan if len(i)<1 else int(i) for i in [i for i in values][0]])
        if len(vector[vector>0]>=180):
            count1=len(vector[vector>0])
            vector = vector.interpolate(method='values') #Do interpolation first
            vector = vector.replace(np.nan,vector.mean()) #Replace remaining NAN with mean of vector now
            count2=len(vector[vector>0])


            logfile.write('REDUCER1:'+station+'\t'+str(count1)+'\t'+str(count2)+'\n')
            yield key_next, dumps(vector)
        
    def outerproduct(self,station,values):
        """
        Key for this step is STATION+YEAR
        Key for next step is STATION
        """
        s,y=station.split(' ')
        key_next = str(self.buckets[s])
        """
        1.Concatenate the TMAX and TMIN vectors
        2.Find outer product for each
        """
        vector=[loads(i) for i in values]
        if len(vector)==2:
            X=vector[0].append(vector[1])
            X=np.matrix([i for i in X])
            outerprod = np.outer(X,X)
            logfile.write('REDUCER2:'+station+'\t'+str(len(vector))+'\t'+str(outerprod.shape)+'\n')
            yield key_next,dumps(outerprod)
            
    def sum_outer_product(self,bucket,values):
        vector=[loads(i) for i in values]
        sum_of_all_prods = sum(v for v in vector)
        logfile.write('REDUCER3:'+bucket+'\t'+str(len(vector))+'\t'+str(type(sum_of_all_prods))+'\t'+str(sum_of_all_prods)+'\n')
        yield bucket,dumps(sum_of_all_prods)
        
    def steps(self):
        return [self.mr(mapper_pre_filter='grep "TMAX\|TMIN"',                        
                        mapper=self.mapper_1,
                        reducer=self.reducer_1),
                self.mr(reducer_init=self.get_buckets,
                        reducer=self.outerproduct),
                self.mr(reducer=self.sum_outer_product)
                ]

if __name__ == '__main__':
    PCA.run()
```
