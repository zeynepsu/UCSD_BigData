### Valid Measurement

First, we have to define a valid measurement, if one station has the following condition, we can increase the weight of the station by 1 or we can say that the station has one more valid measurment.

1) The year A which has both 'TMIN' measurement and 'TMAX' measurement;

2) In year A, there are over half of year days, which have both 'TMIN' and 'TMAX' nonan data

###### Basic Algorithm

We use Mapreduce to statistic the weight of each station

The algorithm is following,

mapper1:
    # find all measurement with 'TMAX' and 'TMIN'
    data => [station, year], [data]
reducer1:
    # check each station on each year, if it statisfied the above two 
    # condition, yield 1 weight for that station 
    [station, year], vector[data] => station, 1/0
reducer2:
    # output the overall weight of each station
    station, 1/0 => station, weight

```python
%%writefile valid_measurement.py
from __future__ import division
from mrjob.job import MRJob

class ValidMeasurement(MRJob):
    def mapper1(self, _, line):
        elements = line.split(',')
        if elements[1] == 'TMIN' or elements[1] == 'TMAX':
            yield (elements[0], elements[2]), elements[3:]

    def reducer1(self, pair, vectors):
        vectors = list(vectors)
        if len(vectors) < 2:
            yield pair[0], 0
        else:
            day_count = 0 
            for i in range(len(vectors[0])):
                if vectors[0][i] != '' and vectors[1][i] != '': 
                    day_count += 1
            if day_count / len(vectors[0]) < 0.5:
                yield pair[0], 0
            else:
                yield pair[0], 1
    
    def reducer2(self, station, is_valid):
        weight = sum(is_valid)
        if weight != 0:
            yield station, weight
    
    def steps(self):
        return [self.mr(mapper=self.mapper1, reducer=self.reducer1),
                self.mr(reducer=self.reducer2)]

if __name__ == '__main__':
    ValidMeasurement.run()
```

## 2 Data Partition

Here, I will use two methods to partition the data. One is **QuadTree** and another one is **KD-Tree**.

### QuadTree

A quadtree is a tree data structure in which each internal node has exactly four children. Quadtrees are most often used to partition a two-dimensional space by recursively subdividing it into four quadrants or regions. The regions may be square or rectangular, or may have arbitrary shapes. More detail on http://en.wikipedia.org/wiki/Quadtree

1. They decompose space into adaptable cells
2. Each cell (or bucket) has a maximum capacity. When maximum capacity is reached, the bucket splits
3. The tree directory follows the spatial decomposition of the quadtree.

**QuadTree Algorithm**:

    1. set the max data number for one QuadTree
    2. if the size of data <= the max num: create a QuadTree with all data
    3. if the size of data >  the max num 
        3.1 seperate the data into four different Quadrants
        3.2 create four QuadTree Children: nw, ne, se, sw

```python
from __future__ import division
import pandas as pd
from mrjob.job import MRJob
import random

labels = np.arange(10000)
random.shuffle(labels)
regions = labels
count = 0

class QuadTree:
    def __init__(self, stations, depth, bounding_rect, maxIt):
        global count
        self.nw = self.ne = self.se = self.sw = None
        self.maxIt = maxIt
        self.depth = depth
        cum_sum = stations.weight.cumsum()
        if cum_sum.iloc[-1] <= self.maxIt:
            print '-----Leaf-----', cum_sum.iloc[-1], '----Stations----', len(stations)
            Djoined.loc[stations.index,'label']= regions[count]
            #DD.loc[stations.index,'label']= regions[self.depth]
            self.region = regions[count]
            count += 1
            self.leaf = True
            return

        self.bounding_rect = bounding_rect
        l, t, r, b = bounding_rect
        cx = self.cx = ( l + r ) * 0.5 
        cy = self.cy = ( t + b ) * 0.5 

        self.items = []
        nw_items = []
        ne_items = []
        se_items = []
        sw_items = []

        for item in stations.index:
            item_x = stations.loc[item, 'latitude']
            item_y = stations.loc[item, 'longitude']
            
            in_nw = item_x <= cx and item_y >  cy  
            in_ne = item_x >  cx and item_y >= cy
            in_sw = item_x <= cx and item_y <  cy  
            in_se = item_x >  cx and item_y <= cy
            
            if   in_nw: nw_items.append(item)
            elif in_ne: ne_items.append(item)
            elif in_sw: sw_items.append(item)
            elif in_se: se_items.append(item)

        if nw_items:
            self.nw = QuadTree(stations.loc[nw_items], depth+1, (l, t, cx, cy),maxIt)
        if ne_items:
            self.ne = QuadTree(stations.loc[ne_items], depth+1, (cx, t, r, cy),maxIt)
        if se_items:
            self.se = QuadTree(stations.loc[se_items], depth+1, (cx, cy, r, b),maxIt)
        if sw_items:
            self.sw = QuadTree(stations.loc[sw_items], depth+1, (l, cy, cx, b),maxIt)
```

### KD-Tree

A k-d tree (short for k-dimensional tree) is a space-partitioning data structure for organizing points in a k-dimensional space. k-d trees are a useful data structure for several applications, such as searches involving a multidimensional search key (e.g. range searches and nearest neighbor searches). k-d trees are a special case of binary space partitioning trees. More detail on http://en.wikipedia.org/wiki/K-d_tree

```python
from __future__ import division
import pandas as pd
from mrjob.job import MRJob
import random

labels = np.arange(10000)
random.shuffle(labels)
regions = labels
count = 0

class Node:
    def __init__(self,la,lo,il,n,l,r,ii=[]):
        self.latitude  = la
        self.longitude = lo
        self.isleaf = il
        self.num = n
        self.left = l
        self.right = r
        self.item = ii
    def __str__(self):
        print 'Is Leaf: ', self.isleaf
        print 'latitude:', self.latitude
        print 'longitude', self.longitude
        if self.left != None: print 'left\n', self.left
        if self.right != None: print 'right\n', self.right

class KDtree:
    def __init__(self, bincap, stations):
        self.bincap = bincap
        self.root = self.creat_tree(stations)
    
    def creat_tree(self, stations, depth=0):
        global count 
        cum_sum = stations.weight.cumsum()
        if cum_sum.iloc[-1] <= self.bincap:
            print '-----Leaf-----', cum_sum.iloc[-1], '----Stations----', len(stations)
            count += 1
            Djoined.loc[stations.index,'label']=regions[count]
            return Node(None,None,True,regions[count],None,None,stations.index)
                      
        isLa = False
        isLo = False
        if depth % 2 == 0:
            isLa = True
            stations.sort('latitude',ascending=True,inplace=True)
        else:
            isLo = True
            stations.sort('longitude',ascending=True,inplace=True)

        median = 0 
        cum_sum = stations.weight.cumsum()
        for i in range(len(cum_sum)):
            if cum_sum.iloc[i] >= cum_sum.iloc[-1] / 2:
                median = i 
                break    
        if median == len(cum_sum)-1:
            median -= 1
            
        if isLa == True:    
            return Node(stations.iloc[median]['latitude']-0.00001,
                        None, False, None,
                        self.creat_tree(stations[:median+1],depth+1),
                        self.creat_tree(stations[median+1:],depth+1))
        elif isLo == True:
            return Node(stations.iloc[median]['longitude']-0.00001,
                        None, False, None,
                        self.creat_tree(stations[:median+1],depth+1),
                        self.creat_tree(stations[median+1:],depth+1))
    def trace_back(self, node):
        if node.isleaf == False:
            left_info = self.trace_back(node.left)
            right_info = self.trace_back(node.right)
            [C_l, N_l] = left_info
            [C_r, N_r] = right_info
            C = np.multiply(C_l, N_1) + np.multiply(C_r, N_r)
            N = N_l + N_r
            node.info = [C, N]
            return node.info
        else:
            return node.info
```

```python
%%writefile pca_test.py
#!/usr/bin/python
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
import re,pickle,base64,zlib
from mrjob.job import MRJob
from sys import stderr
import pandas as pd
import numpy as np
import sklearn as sk
import gzip
import pickle

f = gzip.open('station_hash.pkl.gz', 'rb' )
pickleFile = pickle.Unpickler(f)
hashtable = pickleFile.load()
f.close()

class PCA_weather(MRJob): 

    #This loads the lookup file into a field on the object
    def load_data(self):
        global hashtable
        f = gzip.open('station_hash.pkl.gz', 'rb' )
        pickleFile = pickle.Unpickler(f)
        hashtable = pickleFile.load()
        f.close()
    
    def mapper_pre(self, _, line):
        global hashtable
        elements = line.split(',')
        if elements[1] == 'TMAX' or elements[1] == 'TMIN':
            region = hashtable.get(elements[0])
            if region != None:
                yield (elements[0], elements[2], region), (elements[1], elements[3:])
                
    def reducer_pre(self, station_pair, weathers):
        weather_max = []
        weather_min = []
        valid_count = 0
        data_count = 0
        for weather in weathers:
            if weather[0] == 'TMAX': weather_max = weather[1]
            if weather[0] == 'TMIN': weather_min = weather[1]
            data_count += 1
        if data_count == 2 and len(weather_max) == 365 and len(weather_min) == 365:
            for i in range(365):
                if weather_max[i] != '' and weather_min[i] != '':
                    valid_count += 1
            if valid_count >= 180:
                weather_data = weather_max + weather_min
                yield station_pair[2], weather_data
            
    def reducer_pca(self, region, matrixs):
        M = pd.DataFrame(matrixs)
        M[M == ''] = float('NaN')
        M = M.astype(float)
        M = M.transpose()
        (columns,rows)= np.shape(M)
        Mean = np.mean(M, axis=1).values
        C=np.zeros([columns,columns])  
        N=np.zeros([columns,columns])

        for i in range(rows):
            row = M.iloc[:,i] - Mean
            outer = np.outer(row,row)
            valid = np.isnan(outer) == False
            C[valid] = C[valid]+ outer[valid]
            N[valid] = N[valid]+ 1
            
        valid_outer = np.multiply(1 - np.isnan(N),N>0)
        cov = np.divide(C,N)
        cov = np.multiply(cov, valid_outer)
        U, D, V = np.linalg.svd(cov)
        cum_sum = np.cumsum(D[:])/np.sum(D)
        for i in range(len(cum_sum)):
            if cum_sum[i] >= 0.99:
                ind = i 
                break
        yield region, ind
    
    def steps(self):
        return [self.mr(mapper = self.mapper_pre, reducer=self.reducer_pre),
                self.mr(reducer = self.reducer_pca)]
        
if __name__ == '__main__':
    PCA_weather.run()
```

## 4 Relabel the regions

Here, we use **Geohash** to econde each region, more detail can be found on http://en.wikipedia.org/wiki/Geohash

          A             A => 0
        /   \
       B     C          B => 00    C => 01
     /   \ /   \
    D    E F    G       D => 000   E => 001  F => 010  G =>011

In this way, when we want to find the neighbor, we just check the prefix of each node.

```python
def relabel_tree(root, label):
    is_leaf_parent = False
    if root.left.isleaf == True:
        is_leaf_parent = True
        Djoined.loc[root.left.item,'label'] = label+'0'
    if root.right.isleaf == True:
        is_leaf_parent = True
        Djoined.loc[root.right.item,'label'] = label+'1'
    if is_leaf_parent == False:
        relabel_tree(root.left, label+'0')
        relabel_tree(root.right, label+'1')
    else:
        return
```

## 5 Merge the Stations

Here, we merge two stations based on the criteria given by http://seed.ucsd.edu/mediawiki/index.php/BigDataHW2

For two node, A and B, on the same level, for their parent node C, if we have 
$$n_A * k_A + (k_A+1)\cdot(2*365) + n_B * k_B + (k_B+1)\cdot(2*365) > n_{C} * k_{C} + (k_{C}+1)\cdot(2*365)$$
we can merge two node.

If child node A have two region in it, we will skip that level and search other possible merge.


### Merge Station using mapreduce

In order to merge the leaves, we can find sibs by checking the prefix of each node

Such as for node A with '0101101', node B with '0101100' and node C with '1000100'

we cut off the last character of each node, and we have,

A => '010110'     B => '010110'      C => '100010'

Then we can merge A and B, which have the same prefix

```python
%%writefile pca_merge.py
#!/usr/bin/python
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
import re,pickle,base64,zlib
from mrjob.job import MRJob
from sys import stderr
import pandas as pd
import numpy as np
import sklearn as sk
import gzip
import pickle

f = gzip.open('station_hash_re.pkl.gz', 'rb' )
pickleFile = pickle.Unpickler(f)
hashtable = pickleFile.load()
f.close()

class PCA_merge_weather(MRJob): 

    #This loads the lookup file into a field on the object
    def load_data(self):
        global hashtable
        f = gzip.open('station_hash.pkl.gz', 'rb' )
        pickleFile = pickle.Unpickler(f)
        hashtable = pickleFile.load()
        f.close()
    
    def mapper_pre(self, _, line):
        global hashtable
        elements = line.split(',')
        if elements[1] == 'TMAX' or elements[1] == 'TMIN':
            region = hashtable.get(elements[0])
            if region != None:
                yield (elements[0], elements[2], region), (elements[1], elements[3:])
                
    def reducer_pre(self, station_pair, weathers):
        weather_max = []
        weather_min = []
        valid_count = 0
        data_count = 0
        for weather in weathers:
            if weather[0] == 'TMAX': weather_max = weather[1]
            if weather[0] == 'TMIN': weather_min = weather[1]
            data_count += 1
        if data_count == 2 and len(weather_max) == 365 and len(weather_min) == 365:
            for i in range(365):
                if weather_max[i] != '' and weather_min[i] != '':
                    valid_count += 1
            if valid_count >= 180:
                weather_data = weather_max + weather_min
                region = station_pair[2]
                yield region[:-1], weather_data
            
    def reducer_pca(self, region, matrixs):
        M = pd.DataFrame(matrixs)
        M[M == ''] = float('NaN')
        M = M.astype(float)
        M = M.transpose()
        (columns,rows)= np.shape(M)
        Mean = np.mean(M, axis=1).values
        C=np.zeros([columns,columns])  
        N=np.zeros([columns,columns])

        for i in range(rows):
            row = M.iloc[:,i] - Mean
            outer = np.outer(row,row)
            valid = np.isnan(outer) == False
            C[valid] = C[valid]+ outer[valid]
            N[valid] = N[valid]+ 1
            
        valid_outer = np.multiply(1 - np.isnan(N),N>0)
        cov = np.divide(C,N)
        cov = np.multiply(cov, valid_outer)
        U, D, V = np.linalg.svd(cov)
        cum_sum = np.cumsum(D[:])/np.sum(D)
        for i in range(len(cum_sum)):
            if cum_sum[i] >= 0.99:
                ind = i 
                break
        yield region, ind
    
    def steps(self):
        return [self.mr(mapper = self.mapper_pre, reducer=self.reducer_pre),
                self.mr(reducer = self.reducer_pca)]
        
if __name__ == '__main__':
    PCA_merge_weather.run()
```
