### (1) Calculate the weights for each station

### We save the weight data in dictionary type and then tranform it into the type of data frame.

```python
data_in=open('Station_Count', 'r')

station_weight={}

for line in data_in.readlines():
    words=line.split('\t')
    station_name=words[0].split(',')
    station_name[0]=station_name[0].replace('"','')
    station_name[0]=station_name[0].replace('[','')
    
    weight=words[1].split(',')
    weight[0]=weight[0].replace('[','')
    weight[1]=weight[1].replace(']','')
    
    if weight[0] >= 350 and weight[1] >= 350:
        if station_weight.get(station_name[0]):
            station_weight[station_name[0]] = station_weight[station_name[0]] + int(weight[0]) + int(weight[1])
        else:
            station_weight[station_name[0]] = int(weight[0]) + int(weight[1])

print station_weight
```

### (2) Use KD-tree to do the partition based on the weights

### In each step, we find the medium point on either latitude or longitude to do the partition as KD-tree works. The process is done recursively until we reach the given level, for example 10 levels/1024 parts. children_df stores the partition information (partition id) for each station and children_region stores the border information for each partition. One partition id correspondes to one partition area/region.

```python
import pandas as pd

class KD_Partition(object):

    def __init__(self, df, levels):
        self.df = df
        self.levels = levels
        
        
    def get_id(self, bits):
        return sum(bit*(2**i) for bit, i in zip(bits, range(len(bits))))

    
    def execute(self, df, level, partition='latitude', N=90, S=-90, W=-180, E=180, bits=[]):
        if level <= 0:
            id = self.get_id(bits)
            df['id'] = id
            region = {'id': id, 'N': N, 'S': S, 'W': W, 'E': E, 'weight': df.weight.sum()}
            return df, region
        
        children_df = []
        children_region = []
        df = df.sort(partition)
        df['cumulative_weight'] = df.weight.cumsum()
        threshold = df.weight.sum()/2
        tree_id = 'tree' + str(level)
        df[tree_id] = df.cumulative_weight > threshold
        del df['cumulative_weight']
        half_df = df.groupby(tree_id)

        below_df = pd.DataFrame()
        above_df = pd.DataFrame()
        for flag, half in half_df:
            if flag: above_df = half
            else: below_df = half

        min_Value = above_df[partition].min()
        max_Value = below_df[partition].max()
        border = (max_Value + min_Value)/2

        for flag, child in half_df:
            if partition == 'latitude':
                if flag: S = border
                else: N = border
                child_df, child_region = self.execute(child, level-1, 'longitude', N, S, W, E, [flag]+bits)
            else:
                if flag: W = border
                else: E = border
                child_df, child_region = self.execute(child, level-1, 'latitude', N, S, W, E, [flag]+bits)
        
            children_df.append(child_df)
            children_region.append(child_region)

        return pd.concat(children_df), children_region
```

```python
kd_tree = KD_Partition(Djoined, 10)
kd_df, kd_region = kd_tree.execute(Djoined, 10)
```

### (3) Draw the map to show the partition results.

### For each partition id/part, we give different color to show. Since there are most stations in the US, the US also has the most partition parts.

```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt

lonmin=-180;lonmax=180;latsmin=-80;latsmax=80;
plt.figure(figsize=(15,10),dpi=300)
m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax, llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
m.drawcoastlines()
m.fillcontinents(color='coral',lake_color='aqua')
parallels = np.arange(-80,81,10.)
meridians = np.arange(10.,351.,20.)
m.drawparallels(parallels,labels=[False,True,True,False])
m.drawmeridians(meridians,labels=[True,False,False,True])
m.drawmapboundary(fill_color='aqua')
x,y = m(kd_df['longitude'].tolist(),kd_df['latitude'].tolist())
m.colorbar(m.scatter(x,y,c=kd_df['id'].tolist(),marker='.',zorder=15, s=5,lw=0),location='bottom',pad="5%")

plt.title('# of measurements')
plt.show()
```
