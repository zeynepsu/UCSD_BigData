## 2. Space partitioning of world map using k-d tree

From the plots in the previous section, we have got a good sense of how the temperature measurements are distributed across the globe. We have found that the measurements are very inhomogeneously distributed. Our goal now is to partition the world map into regions with a similar number of measurements.

### 2.1 Generating a k-d tree

One possible approach is to iteratively split the regions into two partitions, such that each partition has about the same number of measurements. In other words, we will split each region at the median of the cumulative measurement distribution. As we have two coordinates, namely longitude and latitude, we will alternate between these two coordinates in each partitioning step.

The just described algorithm can be implemented as a k-d tree, where each node has the following form:

```python
from collections import namedtuple
from pprint import pformat

class Node(namedtuple('Node', 'node_id key left_child right_child')):
    def __repr__(self):
        return pformat(tuple(self))
```

In our case, the key consists of the median longitude or latitude (``median``) which will separate the region into two partitions, and the total number of measurements (``total``) contained in the region that is splitted. The key information ``total`` is not strictly required, but will be a nice feature later on for sanity checks.

The function ``generate_kdtree(data,...,threshold)`` generates a k-d tree based on the input ``data``. ``data`` has to be a $N\times 3$ matrix, containing the longitude, latitude and measurement count information as columns vectors (each with $N$ entries). It starts by dividing the region at the median with respect to longitude and will alternate between longitude and latitude as the splitting axis at each following iteration step. Another important input parameter is ``threshold``, which defines the maximum number of measurements a region should have in order from not being partitioned any further.

```python
def generate_kdtree(data, depth=0, nodeid=0, threshold=10000):
    # Stop if under threshold
    total = sum(data[:,2])
    if total<threshold:
        return Node(
            node_id=nodeid,
            key=(None,total,data[:,0],data[:,1]),
            left_child=None,
            right_child=None
        )
    # Stop if only one data point left
    if len(data[:,0])<=1:
        return Node(
            node_id=nodeid,
            key=(None,total),
            left_child=None,
            right_child=None
        )
    # Select axis based on depth so that axis cycles through dimensions
    axis = depth % 2
    # Sort point list and choose median as pivot element
    idx = data[:,axis].argsort()
    data = data[idx] 
    # Compute median
    maxlimit = total/2
    counts = 0
    N = len(data[:,2])
    for j in range(N):
        counts = counts + data[j,2]
        if counts >= maxlimit:      
            break        
            
    if j == 0:
        median = (data[j][axis]+data[j+1][axis])/2
        left_data = data[:1]
        right_data = data[1:]
    else:
        median = (data[j-1][axis]+data[j][axis])/2   
        left_data = data[:j]
        right_data = data[j:]
            
    # Create node and construct subtrees
    return Node(
        node_id=nodeid,
        key=(median,total,j,N-1),
        left_child=generate_kdtree(left_data,depth+1,2*nodeid+1,threshold),
        right_child=generate_kdtree(right_data,depth+1,2*nodeid+2,threshold)
    )
```

Now, let us generate a k-d tree for the measurement count ``N_TMIN365`` with a threshold of 100,000 measurements. For that, we first need to import the data and measurement table ``dfComplete`` created in the previous part.

### 2.2 Partitioning the weather stations

Given the dataframe ``dfComplete`` we created in the first section, we want to add a "partition ID" to each station depending on in which partition the station is located.

```python
dfComplete.head()
```

For this purpose, I defined the function ``find_partition`` that - for a given station - will walk down the k-d tree based on the station's longitude and latitude information until it ends up at a leaf. The leaf's node ID will then be used as the partition ID. In addition, it will return the boundaries of each partition. This will be useful, if we want to merge neighboring partitions later on.

```python
def find_partition(location,kdtree,axis=0,bounds=[-180,180,-90,90]):
    if location[axis] < kdtree[1][0]:
        if kdtree[2] is None:
            # return nodeid
            return kdtree[0], bounds
        else:
            # going to left child
            bounds[axis*2+1]=kdtree[1][0]
            return find_partition(location,kdtree[2],(axis+1)%2,bounds)
    else:
        if kdtree[3] is None:
            # return nodeid
            return kdtree[0], bounds
        else:
            # going to right child
            bounds[axis*2]=kdtree[1][0]
            return find_partition(location,kdtree[3],(axis+1)%2,bounds)         
```

```python
partitionID = np.zeros([len(lons),1])
partitionBounds = np.zeros([len(lons),4])
location = np.zeros([len(lons),2])
location[:,0] = lons
location[:,1] = lats

for j in range(len(lons)):
    partitionID[j],partitionBounds[j,:] = find_partition(location[j,:],kdtree)
```

Store table with added partition ID information into a new dataframe named ``dfFinal``.

```python
dfFinal = dfComplete.copy()
dfFinal.insert(1,'lon_max',partitionBounds[:,1])
dfFinal.insert(1,'lon_min',partitionBounds[:,0])
dfFinal.insert(1,'lat_max',partitionBounds[:,3])
dfFinal.insert(1,'lat_min',partitionBounds[:,2])
dfFinal.insert(1,'partitionID',partitionID)
```

```python
dfPartition.ix[:,'N_TMIN365'].hist(bins=100)
```

We can see that most partitions contain less then 100,000 measurements, which was the threshold set in the beginning. However, we also see that there are some "unlucky" splits in the sense that some partitions contain relatively few measurements. 

Moreover, a few partions seem to have more than 100,000 measurements. This is because there are cases, where a region was split exactly through a station. This can happen, if the median is positioned between two stations with same longitude (or latitude), such that the median is exactly the longitude (or latitude) of the two stations. In this case, the ``find_partition`` algorithm assigns both stations to either region, although they were placed in separate regions when generating the k-d tree. 

I'm sure there is a solution to fix this issue, but for now and focusing on the following PCA analysis let us accept the current partition structure.
