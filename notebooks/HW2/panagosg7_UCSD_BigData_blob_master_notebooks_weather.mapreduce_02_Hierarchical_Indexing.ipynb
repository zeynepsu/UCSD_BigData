#Hierarchical Indexing (HI) Structure

This class holds and manipulates information regarding:
 - **stations** (coordinates, weights)
 - partitioning of the world in **rectangle** areas
 - partitioning of the world in **groups** of rectangle areas
 
In more detail, the **constructor** of this class takes as input:
 - the **station** dataframe that we computed in `01_Station_Weights.ipynb`, and
 - the **level** of the Hierarchical Indexing structure, which is a *k-d-tree*
 
 
####Map partitioning
The goal of the hierarchical partitioning of the data is that siblings (in our case
**rectangle** areas) are close
to each other and leaves have a similar "weight". 

Method `run` does the partitioning. It takes as input the boundaries of an area and 
splits it either 
horizontally or vertically (this choice alternates as we move deeper in the structure).
The point at which the split happens is such that the cumulative weight of the 
stations in either side is *about* the same.

This process results in the partitioning of the world in rectangles of about the same
total weight.

Besides the partitioning of the map in rectangles, there is also a partitioning 
into **groups**. Initially these two partionings are exactly the same. However, as 
our analysis procedes we might be able to merge two **neighboring** groups into a 
single one. For this reason we need to keep information for group that contains 
among others: 
 - the rectangles that comprise it, and
 - the neighboring groups


####Merging of groups
One of the main operations that can be performed on the HI structure is merging of 
groups. This operation is only possible between beighboring groups. Two groups are 
*neighboring* if they share some portion of their boundaries. After merging, we need 
to update the dataframe that keeps track of the station information, as new station
ids are introduced and neighbors might change.

##Local Properties Description

* **`_stations`**: a dataframe with the original station information (immutable)  
* **`_levels`**: the number of levels of the *k-d-tree*
* **`_grouped_stations`**: dataframe containing station information, including the group
    that the station belong to at all times. This structure gets updated at merging of
    groups.
* **`_rectangles`**: dataframe indexed by the `id` for each rectangle of the original 
    partitioning of the global map. This structure remains invariant after initialization.
* **`_groups`**: dictionary indexed by the group `id`, containing information about each 
    group (neighbors, weight, included rectangles). Gets updated at group merging.
* **`_computed`**: boolean value designatingif the original partition is done.
* **`_neighbors`**: boolean square matrix indexed by rectangle ids at rows and columns. 
    A cell in this matrix is `True` if the indexing rectangles are neighboring, `False` 
    otherwise.
* **`_gid_counter`**: counter used to provide "fresh" group ids at merging


##Main Methods Descriptions
* **`get_grouped_stations()`**: returns the station dataframe of the **current** grouping
    of stations (i.e. taking into account previous merges)
* **`get_neighbors_of_group(i)`**: returns the **current** neighbors of group `i`.
* **`groups_are_neighbors(i,j)`**: returns `True` if groups `i` and `j` are currently 
    neighbors
* **`merge_groups(i,j)`**: perform a merge of groups `i` and `j`. These groups need to be
    neighboring otherwise nothing will happen. This method takes care of all the accounting 
    outlined above, and returns the id of the new group that gets created.
* **`run`**: perform the partitioning based on the original "weighted" station dataframe.
    This does not need to be called from outside the class as it will be called internally 
    if any of the above structures has not been initiallized.
* **`__run__(df, level, coord, west, east, south, north, path)`**: this method recursively
    performs the partitioning. It is only meant to be called by `run`. The coordinates `west`,
    `east`, `north` and `south` designate the area to be partitioned. `df` is a dataframe 
    that **only** contains stations in this area. `level` is the level on the *k-d-tree* we 
    are at and `path` is a summary of the conditions that hold at the current node of the tree, 
    i.e. describes the path we need to follow until this point. At leaf level, this path is used
    to compute a unique id for each rectangle. This function partitions nodes into two areas:
    a `FALSE` and a `TRUE`, by adding a vertical or horizontal boundary, based on the value
    if `coords`.
   
      if `coord == 'longitude'`:
                               north
                      ______________________________
                     |FALSE                 |   TRUE|   (90)
                     |               new -> |       |
                west |             boundary |       | east
                     |                      |       | 
                     |______________________|_______|   (-90) 
                   (-180)       south             (180)   
                   

      if `coord = 'latitude'`:
                               north
                      _______________________
                     |TRUE           new     |  (90)
                     |             boundary  |
                west |                |      | east
                     |________________v______|
                     |FALSE                  |
                     |                       |
                     |_______________________|  (-90) 
                   (-180)       south      (180)   

    
* **`plot_map(title)`**: Plot a map of the world such that:
   - it is paritioned into the rectangles that were created at initialization.
   - each rectangle is colored randomly, but rectangles that belong to the same 
     group (based on the **current** state of the grouping) are of the same color. Of course 
     these can only be neighboring rectangles.

```python
%%writefile hierarchical_indexing.py
import pandas as pd
import numpy as np
import sklearn as sk
print 'pandas version: ',pd.__version__
print 'numpy version:',np.__version__
print 'sklearn version:',sk.__version__


class HierarchicalIndexing:
    
    # Pandas.DataFrame indexed by station ID, containing at least the cilumns:
    # * weight
    # * longitude
    # * latitude
    _stations         = None
    
    # int: levels of hierarchy tree
    _levels           = 0
    
    # a new Pandas.DataFrame containing station information including the 'group_id' 
    _grouped_stations = None
    
    # Panda.DataFrame indexed by 'rect_id'
    _rectangles       = None
    
    # Dictionary indexed by 'group_id' with properties:
    #  * weight    : int 
    #  * rects     : array of rect_id
    #  * neighbors : array of group_id
    # 
    # TODO: optimize using bitvector instead of array for 'neighbors'
    _groups           = None
    
    # boolean: True if the analysis has run, False otherwise
    _computed         = False
    
    # np.ndarray with dimensions: (2**levels, 2**levels)
    _neighbors        = None
    
    # a counter for fresh gid's
    _gid_counter      = 0
    
    
    
    def __init__(self, df, levels):
        """
        Initializer

        Parameters
        ----------
        df    : Pandas.DataFrame
                Indexed by station ID
                Required fields:
                 * weight
                 * longitude
                 * latitude

        level : int
                  Level of iteration - reasonable values are > 0 
        """        
        
        # set level
        self._levels = levels
        
        # keep a copy of the original input dataframe
        self._stations = df


    def get_groups(self):
        if not self._computed:
            self.run()
        return self._groups

    
    def get_stations(self):
        if not self._computed:
            self.run()
        return self._stations

    
    
    def get_grouped_stations(self):
        if not self._computed:
            self.run()
        return self._grouped_stations

    
    def next_c(self,c):
        if   c == 'longitude':
            return 'latitude'
        elif c == 'latitude':
            return 'longitude'
        else:
            raise Exception('long/lat')

            
    def level_str(self,i):
        return 'g' + str(i)
    
    
    def fresh_gid(self):
        num = self._gid_counter
        self._gid_counter = num + 1
        return num
        
        
    def get_rectangles(self):
        if not self._computed:
            self.run()
        return self._rectangles
        
    
    
    ## Rectangle Neighbors 
    
    def __overlap__(self,a1,b1,a2,b2):
        assert(a1 < b1)
        assert(a2 < b2)
        if b1 < a2 or b2 < a1: 
            return False
        return True
    
    
    def eq(self,a,b):
        return np.absolute(a - b) < 0.01
    
    
    def __neighboring__(self,i,j):
        """
        Returns True if rectangle `i` is neighboring with rectangel `j`, 
        otherwise returns False.
        
        TODO: wrap around the globe
        
        """
        gi = self._rectangles.loc[i]
        gj = self._rectangles.loc[j]
        
        # top-bottom
        if self.eq(gj.north, gi.south) or self.eq(gi.north, gj.south):
            if self.__overlap__(gi.west, gi.east, gj.west, gj.east):
                return True
        
        # left-right
        if self.eq(gj.east, gi.west) or self.eq(gi.east, gj.west):
            if self.__overlap__(gi.south, gi.north, gj.south, gj.north):
                return True
        
        return False
        
    
    def __compute_neighbors__(self):
        if not self._computed:
            self.run()
                
        rect_no = len(self._rectangles)
        
        self._neighbors = np.zeros((rect_no,rect_no), dtype=np.bool)
        
        for i in range(0,rect_no):
            for j in range(i,rect_no):
                if self.__neighboring__(i,j):
                    ## Add edge
                    self._neighbors[i][j] = True
                    self._neighbors[j][i] = True
                    
                    
    def get_neighbor_rectangles(self):
        if not self._computed:
            self.run()            
        return self._neighbors
        
    
    def get_neighbors_of_rectangle(self,n):
        """       
        For a rectangle with rect_id `n`, returns a list of the neighboring rectangles.
        """
        n_rect = self.get_neighbor_rectangles()
        rect_no = len(self._rectangles)
        for j in range(0,rect_no):
            if n_rect[n][j]:
                yield j

                
    def rectangles_are_neighbors(self,i,j):
        """       
        Given two rectangles `i` and `j`, returns True if these are neighboring,
        False otherwise.
        """
        narr = self.get_neighbor_rectangles()
        return narr[i][j]            

        
    
    ## Group Neighbors
    
    def get_neighbors_of_group(self,n):
        """       
        For a group with group_id `n`, returns a list of the neighboring groups.
        """
        groups = self.get_groups()
        return groups[n]['neighbors']
        
                
    def groups_are_neighbors(self,i,j):
        """       
        Given two groups `i` and `j`, returns True if these are neighboring,
        False otherwise.
        """
        return j in self.get_neighbors_of_group(i)
    
    
    
    
    # Operations on groups
    def merge_groups(self,i,j):
        """
        Returns the group_id of the new group that gets created when merging 
        groups `i` and `j`. If these groups cannot be merged (for instance if one
        of them has been merged already) then return -1.
        """        
        groups = self.get_groups()            
                
        if (i in groups) and (j in groups) and self.groups_are_neighbors(i,j):
            gi = groups[i]
            gj = groups[j]
            
            ri = gi['rects']
            rj = gj['rects']
            
            ni = gi['neighbors']
            nj = gj['neighbors']
            
            new_g_id = self.fresh_gid()
            new_rects = ri + list(set(rj) - set(ri))            
            new_neighbors = list((set(ni) | set(nj)) - set([i,j]))
            new_weight = gi['weight'] + gj['weight']
            
            groups[new_g_id] = { 
                'weight'   : new_weight,
                'rects'    : new_rects,
                'neighbors': new_neighbors
            }
                        
            # update the properties of the neighboring nodes
            for n in new_neighbors:
                gn = groups[n]
                gn['neighbors'] = [new_g_id] + [ n for n in gn['neighbors'] 
                                                   if n != i and n != j ]
                  
            # delete orig groups
            del groups[i]
            del groups[j]
        
            # update the _grouped_stations entry these groups
            gs = self.get_grouped_stations()
            gs.loc[gs.group_id == i, 'group_id'] = new_g_id
            gs.loc[gs.group_id == j, 'group_id'] = new_g_id
            
            return new_g_id
        
        else:
            return -1

    
    
    #
    # Top-level runner 
    #
    def run(self):        
        if not self._computed:
            self._computed = True            

            # make a deep copy every time `run` is called
            df = self._stations.copy(deep=True)
            
            # run the analysis
            (self._grouped_stations, rect_list) = self.__run__(df, self._levels)            
            
            columns = ['rect_id', 'west', 'east', 'south', 'north', 'weight']
            self._rectangles = pd.DataFrame(rect_list, columns=columns).set_index('rect_id')
            self.__compute_neighbors__()            
            
            # Build the groups dictionary - at this point it coincides with the rectangles
            self._groups = {}
            for rect_id, row in self._rectangles.iterrows():
                self._groups[rect_id] = {
                    'weight'   : row.weight,
                    'rects'    : [rect_id],
                    'neighbors': [ i for i in self.get_neighbors_of_rectangle(rect_id)]
                }
                
            # initialize the gid counter
            self._gid_counter = len(self._groups)

            
    def path_to_key(self, a): 
        return sum([ x * (2**i) for (x,i) in zip(a, range(0, len(a)))])
            
        
    def __run__(self, df, level, coord='latitude', west=-180, east=180, south=-90, north=90, path=[]):
        """
        __run__

        Parameters
        ----------
        df      : Pandas.DataFrame
                  
        level   : int
                  Level of iteration

        coord   : str ('longitide' or 'latitude')
                  The first time this is run it really should be 'latitude' so that we 
                  establish the main separation of northern and southern hemisphere.

        west    : int
                  the western boundary of the region

        east    : int
                  the eastern boundary of the region

        south   : int
                  the southern boundary of the region

        north   : int
                  the northern boundary of the region
                  
        path    : list of boolean
                  the branches taken so far in the hierarchy index tree

        Returns
        -------
        A new dataframe containing:
          * A dataframe with the original stations extended with columns 
            g1, g2, ... that altogether denote the group that each station belongs to.
          * A dictionary with the properties of each leaf node in the tree 
            (i.e. rectangle in the map)
        """
        
        if level <= 0:
            # we have reached a leaf:
            # * compute the rect id
            # * update the stores
            rect_id = self.path_to_key(path)
            rect = { 'rect_id' : rect_id,
                     'west' : west,
                     'east' : east,
                     'south': south,
                     'north': north,
                     'weight' : df['weight'].sum()
                 }
            
            # at this point group_id coincides with rect_id
            df['group_id'] = rect_id
            return (df,[rect])

        # Sort by coord
        df = df.sort(coord)

        # Compute a new column with the cummulative count of 
        df['cum_weight'] = df['weight'].cumsum()

        # compute the threshold of weight (the point that designates the meridian)
        threshold = df['weight'].sum() / 2

        # group index column
        gi = self.level_str(level)

        # assign a group number based on weight
        df[gi] = df['cum_weight'] > threshold

        # drop the 'cum_weight' column
        del df['cum_weight']

        # break down into groups based on the grouping column
        gdf = df.groupby([gi])
        
        # ugly way to get the boundary of the partition on the coordinate of interest
        new_boundary = np.average([[group[coord].max() for name, group in gdf if name==False][0], \
                                   [group[coord].min() for name, group in gdf if name==True ][0]])

        # recursively partition the groups
        children_df = []
        children_rects = []


        for name, child in gdf:
            
            new_east = east
            new_west = west
            new_north = north
            new_south = south
            
            if name==False: 
                # cum_weight is smaller than threshold, i.e.
                #   - station is west of the median if it's longitude, so east gets updated
                #   - station is south of the median if it's latitude, so north gets updated
                if coord=='longitude':
                    new_east = new_boundary
                    #print coord, name, west,'<=', new_east,'<=', east
                else:
                    new_north = new_boundary
                    #print coord, name, south,'<=', new_north,'<=', north

            else:            
                # cum_weight is greater than threshold, i.e.
                #   - station is east of the median if it's longitude, so west gets updated
                #   - station is north of the median if it's latitude, so south gets updated
                if coord=='longitude':
                    new_west = new_boundary
                    #print coord, name, west,'<=', new_west,'<=', east

                else:
                    new_south = new_boundary
                    #print coord, name, south,'<=', new_south,'<=', north

            # do the recursive call using updates boundaries
            # ignore_index = True
            (child_df, child_rect) = self.__run__(child,
                                                level-1,
                                                self.next_c(coord),
                                                new_west,new_east,new_south,new_north,
                                                [int(name)]+path)
            children_df.append(child_df)
            children_rects=children_rects + child_rect

        # concatenate children and return
        return (pd.concat(children_df), children_rects)    


    
    # Plot map
    
    def plot_map(self, title='World groups'):

        # http://matplotlib.org/basemap/users/merc.html
        from mpl_toolkits.basemap import Basemap
        import numpy as np
        import random
        import matplotlib.pyplot as plt

        lonmin=-180;lonmax=180;latsmin=-80;latsmax=80;
        plt.figure(figsize=(15,10),dpi=300)
        m = Basemap(projection='merc', 
                    llcrnrlat=latsmin,urcrnrlat=latsmax,\
                    llcrnrlon=lonmin,urcrnrlon=lonmax, \
                    lat_ts=20,
                    resolution='c')  # 'l','i'
        m.drawcoastlines()
        m.fillcontinents(color='#E2E2E2',lake_color='#FFFFFF')
        m.drawmapboundary(fill_color='#FFFFFF')


        rectangles = self.get_rectangles()
        groups = self.get_groups()

        for group_id in groups:
            group = groups[group_id]
            rect_ids = group['rects']
            # use common color for same group
            r = lambda: random.randint(0,255)
            color = '#%02X%02X%02X' % (r(),r(),r())
            for rect_id in rect_ids:
                rect = rectangles.ix[rect_id]
                draw_screen_poly(plt, group_id, rect.west, rect.east, rect.south, rect.north, m, color, len(groups) < 33)

        # plt.legend()
        plt.title(title)
        plt.show()

    
    
### End of class definition
    
    
# Plotting helpers

def draw_screen_poly(plt, gid, west, east, south, north, m, color, text=True):
    """
    Parameters
    ----------
    plt     : the map plot
    gid     : group's id
    west    : west border  
    east    : east border 
    south   : south border
    north   : north border
    m       : basemap
    color   : color of the rectangle
    text    : print group_id tags on each group on the map (default: True)
    
    Effect
    ------
    Renders a `color` colored rectangle on the map
    """
    def fix(x):
        if x >  80: return 80.0
        if x < -80: return -80.0
        return x
    
    lons = [west, east, east, west]
    lats = map(fix, [south, south, north, north])
    x, y = m( lons, lats )
    xy = zip(x,y)
    from matplotlib.patches import Polygon

    
    poly = Polygon( xy, facecolor=color, alpha=0.4, label = str(gid) )

    
    if text:
        # center    
        clon = (west + east) / 2
        clat = fix((north + south) / 2)
        cx,cy = m([clon],[clat])
        plt.text(cx[0], cy[0], str(gid), 
                 horizontalalignment='center', 
                 verticalalignment='center',
                 bbox=dict(facecolor='white', alpha=1),
                 size =15, fontweight='bold')  
    
    
    plt.gca().add_patch(poly)

    
# Taken from here: http://bensouthgate.com/p/12_3_13.php

def hex_to_RGB(self, hex):
    ''' "#FFFFFF" -> [255,255,255] '''
    # Pass 16 to the integer function for change of base
    return [int(hex[i:i+2], 16) for i in range(1,6,2)] 


def RGB_to_hex(RGB):
    ''' [255,255,255] -> "#FFFFFF" '''
    # Components need to be integers for hex to make sense
    RGB = [int(x) for x in RGB]
    return "#"+"".join(["0{0:x}".format(v) if v < 16 else 
                        "{0:x}".format(v) for v in RGB])
    


def color_dict(gradient):
    ''' Takes in a list of RGB sub-lists and returns dictionary of 
        colors in RGB and hex form for use in a graphing function 
        defined later on '''
    return {"hex":[RGB_to_hex(RGB) for RGB in gradient],
            "r":[RGB[0] for RGB in gradient],
            "g":[RGB[1] for RGB in gradient],
            "b":[RGB[2] for RGB in gradient]}


def linear_gradient(start_hex, finish_hex="#FFFFFF", n=10):
    ''' returns a gradient list of (n) colors between two hex colors.
        start_hex and finish_hex should be the full six-digit color string, 
        inlcuding the number sign ("#FFFFFF") '''
    # Starting and ending colors in RGB form
    s = hex_to_RGB(start_hex)
    f = hex_to_RGB(finish_hex)
    # Initilize a list of the output colors with the starting color
    RGB_list = [s]
    # Calcuate a color at each evenly spaced value of t from 1 to n
    for t in range(1, n):
        # Interpolate RGB vector for color at the current value of t
        curr_vector = [ int(s[j] + (float(t)/(n-1))*(f[j]-s[j])) 
                        for j in range(3)]
        # Add it to our list of output colors
        RGB_list.append(curr_vector)

    return color_dict(RGB_list)
    
```
