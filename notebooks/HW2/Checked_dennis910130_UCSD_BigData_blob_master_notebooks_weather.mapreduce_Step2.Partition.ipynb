Then we use a binary tree structure to encode all the stations. The code for each station represents the partition it belongs to. We do it the way that Yoav posted on Piazza by iteratively partitioning with longitude or latitude. The median is found by the valid years of stations that we found in step1, so that each part contains approximately the same valid years of data.

```python
import numpy
def find_nearest(matrix,value):
    idx = numpy.argmin(numpy.abs(matrix-value))
    return idx


def encoding(matrix, matrix1,code,is_latitude=True):
    
    if is_latitude:
        index = matrix[:,3] == code
        temp_matrix = matrix[index,:]
        sorted_matrix = temp_matrix[numpy.argsort(temp_matrix[:,0])]
        #print sorted_matrix
        matrix_cum = numpy.asarray(sorted_matrix)
        
        matrix_cum[:,2] = numpy.cumsum(sorted_matrix[:,2])
        #print matrix_cum
        median = matrix_cum[matrix_cum.shape[0]-1,2]/2.0
        idx = find_nearest(matrix_cum[:,2],median)
        temp_median = matrix_cum[idx,0]
        print temp_median
        index1 = index&(matrix[:,0] > temp_median)
        index2 = index&(matrix[:,0] < temp_median)
        
        matrix1[index1,3] = code*2
        matrix1[index2,3] = code*2+1
        print 'done'
        
        
    else:
        index = matrix[:,3] == code
        #temp_median = median(matrix[index,1])
        #print 'longitude median is ', temp_median
        temp_matrix = matrix[index,:]
        sorted_matrix = temp_matrix[numpy.argsort(temp_matrix[:,1])]
        matrix_cum = numpy.asarray(sorted_matrix)
        matrix_cum[:,2] = numpy.cumsum(sorted_matrix[:,2])
        median = matrix_cum[matrix_cum.shape[0]-1,2]/2.0
        idx = find_nearest(matrix_cum[:,2],median)
        temp_median = matrix_cum[idx,1]
        index1 = index&(matrix[:,1] > temp_median)
        index2 = index&(matrix[:,1] < temp_median)
        matrix1[index1,3] = code*2
        matrix1[index2,3] = code*2+1
        print 'done'
        
    return code*2, code*2+1, matrix1
```

Then we partition the world into 64 parts based on the algorithm shown above.

```python

tree = stations.loc[:,['latitude_x','longitude_x','Nyears']].copy()
code = numpy.zeros((tree.shape[0],))
tree['code'] = code
matrix = tree.values
print matrix
matrix1 = matrix.copy()
num_iterations = 6
root = [0]
i = 0
while(i < num_iterations):
    if i % 2 == 0:
        temp = []
        for j in range(len(root)):
            new_code = encoding(matrix,matrix1,root[j])
            temp.append(new_code[0])
            temp.append(new_code[1])
            matrix1 = new_code[2]
        root = temp
        matrix = matrix1.copy()
        print root
    else:
        temp = []
        for j in range(len(root)):
            new_code = encoding(matrix,matrix1,root[j],False)
            temp.append(new_code[0])
            temp.append(new_code[1])
            matrix1 = new_code[2]
        root = temp
        matrix = matrix1.copy()
        print root
    i += 1




```
