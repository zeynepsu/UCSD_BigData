```python
%%writefile mr_weather_parition_data.py
import cPickle
Codes = cPickle.load( open( "station_codes1.pkl", "rb" ) )
#import collections
#Codes = collections.defaultdict(lambda: -1,Codes)
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re,pickle,base64,zlib
from sys import stderr
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA





class MRWeather_p_data(MRJob):
    def mapper1(self, _, line):
        self.increment_counter('MrJob Counters','extract_meas_mapper',1)
        elements=line.split(',')
        if elements[1]=='TMAX' or elements[1]=='TMIN':
            Ndays = sum([e!='' for e in elements[3:]])
            meas = 1
            yield (elements[0],elements[2]), (meas,Ndays,elements[3:])
            
    def reducer1(self, key, meas_set):
        self.increment_counter('MrJob Counters','extract_meas_reducer',1)
        sum1 = 0
        sum2 = 0
        total_meas = []
        for Nmeas,Ndays,meas in meas_set:
            sum1 = sum1 + Nmeas
            sum2 = sum2 + Ndays
            total_meas = total_meas + meas
        if sum1 == 2 and sum2>50:
            yield key, (Codes[key[0]],total_meas)
            
    def mapper2(self, key, key_meas):
        value = []
        for each in key_meas[1]:
            if each == '':
                value.append(np.nan)
            else:
                value.append(float(each))
        s = pd.Series(value)
        s = s.interpolate()
        s = s.fillna(s.mean())
        yield key_meas[0], list(s)
    
    def reducer2(self, key, meas):
        tmp = list(meas)
 
        yield key, tmp
    
    
    

    
    def steps(self):

        return [self.mr(mapper=self.mapper1, reducer=self.reducer1),
                self.mr(mapper=self.mapper2, reducer=self.reducer2)]
if __name__ == '__main__':
    MRWeather_p_data.run()
```

```python
%%writefile mr_weather_PCA_data.py
import cPickle
Codes = cPickle.load( open( "station_codes1.pkl", "rb" ) )
#import collections
#Codes = collections.defaultdict(lambda: -1,Codes)
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re,pickle,base64,zlib
from sys import stderr
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
def my_pca(X,n):
    pca = PCA(n_components=n, whiten=True)
    pca.fit(X)
    sum1 = 0
    for i in range(n):
        sum1 = sum1 + pca.explained_variance_ratio_[i]
        if sum1 > 0.99:#the threshold Yoav set
            break
    return i,pca.components_


class MRWeather_p_data(MRJob):
    def mapper1(self, _, line):
        self.increment_counter('MrJob Counters','extract_meas_mapper',1)
        elements=line.split(',')
        if elements[1]=='TMAX' or elements[1]=='TMIN':
            Ndays = sum([e!='' for e in elements[3:]])
            meas = 1
            yield (elements[0],elements[2]), (meas,Ndays,elements[3:])
            
    def reducer1(self, key, meas_set):
        self.increment_counter('MrJob Counters','extract_meas_reducer',1)
        sum1 = 0
        sum2 = 0
        total_meas = []
        for Nmeas,Ndays,meas in meas_set:
            sum1 = sum1 + Nmeas
            sum2 = sum2 + Ndays
            total_meas = total_meas + meas
        if sum1 == 2 and sum2>50:
            yield key, (Codes[key[0]],total_meas)
            
    def mapper2(self, key, key_meas):
        value = []
        for each in key_meas[1]:
            if each == '':
                value.append(np.nan)
            else:
                value.append(float(each))
        s = pd.Series(value)
        s = s.interpolate()
        s = s.fillna(s.mean())
        yield key_meas[0], list(s)
    
    def reducer2(self, key, meas):
        tmp = list(meas)
 
        yield key, tmp
        
    def reducer3(self,key,data):
        pca_data = np.array(list(data)[0])
        explain_length,eigenvectors = my_pca(pca_data,pca_data.shape[1])
        eigv = list(eigenvectors)
        yield key, [pca_data.shape[0],explain_length]
    
    

    
    def steps(self):

        return [self.mr(mapper=self.mapper1, reducer=self.reducer1),
                self.mr(mapper=self.mapper2, reducer=self.reducer2),
                self.mr(reducer=self.reducer3)]
if __name__ == '__main__':
    MRWeather_p_data.run()
```

From the PCA results, we can see that my partitioning algorithm is proper since each region contains roughly the same amount of data.

For the merge task, since I use a binary tree for partitioning the data, it is super easy to find the neighbors of any leaf. 2*n and 2*n+1 are always neighbors(say [40,41],[2,3],etc). 

In order to see whether two leaves should merge, we can simply change the coding file by dividing all the original codings by 2(31->15,30->15). Then use the new coding to do the PCA. Finally, we decide whether to merge the two adjacent leaves using the MDL criterion.
