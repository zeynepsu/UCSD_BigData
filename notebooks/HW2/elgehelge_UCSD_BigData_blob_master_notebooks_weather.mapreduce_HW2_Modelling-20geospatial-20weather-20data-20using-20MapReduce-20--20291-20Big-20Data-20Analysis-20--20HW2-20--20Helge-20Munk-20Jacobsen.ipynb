#### Objective
We want to model the maximum and minimum temperature around the world using principal component analysis (PCA). But since not all places on earth behave the same way, we want to fit PCAs to several different areas in order to get a better fit. This raises the following problems:
* **How do we avoid partitioning the data uneven, so that some areas will overfit and others will get to simple a model?**
* **How do we make a good partitioning of the data into areas that meteorologically behave the same way?**

#### Approach
1. [Aggregation - MapReduce][measurement_count]<br />Find the distrubution of measurements by counting the number of measurements for each weather station using _MapReduce_ (the approximatly 9 million years of measurements are aggregated into counts for each for the around 85000 weather stations).
2. [Partition - R-tree][rtree]<br />Partition the data into areas of approximately equal number of measurements using an _R-tree_ (resulting in areas of uneven size).
3. [Initial model - PCA][pca]<br />Fit models on each area using _principal component analysis (PCA)_.
4. [Merge models - MDL][mdl]<br />Merge areas with similar meteorologically behaviour together using the concept of _minimum description lenth (MDL)_.

### The MapReduce job

In order to count the measurements for each station in the map-reduce framework we need (for each line):

**Mapper:**
The key will be the station ID and the value is the total number of measurements. We only emit a key-value pair if the measurement type is the type of interest. We could also emit a key-value pair for each single measurement, but this would cause extra work for the reducers.

**Reducer:**
Aggregate all measurement counts.

**Combiner**
We don't need a combiner, but if a mapper encounters multiple lines with same station (this happens since each year gets it own line) we could add a combiner that sums all the measurement counts before sending the data to the reducers.

Using the `%%writefile` command we can write the following MRjob Python class to a file.

````python
%%writefile count_TMIN_TMAX.py
#!/usr/bin/python
"""
count the number of measurements of each type
"""
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr

class count_TMIN_TMAX(MRJob):

    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper',1)
            # Parse the line
            elements=line.split(',')
            station = elements[0]
            measure_type = elements[1]
            year = elements[2]
            measurements = elements[3:]
            no_of_measurements = sum([1 for m in measurements if m != ""])
            
            # Emit the key-value pair
            if (measure_type == 'TMIN' or measure_type == 'TMAX') and station != 'station': #filter out the unwanted measurements and the header line
                self.increment_counter('MrJob Counters','usefull lines',1)
                yield (station, no_of_measurements)
                
        except Exception, e:
            stderr.write('Error in line:\n'+line)
            stderr.write(e.message)
            self.increment_counter('MrJob Counters','mapper-error',1)
            yield (('error','mapper', str(e)), 1)
            
    def reducer(self, station, counts):
        try:
            if station[0] == 'error':
                yield(station,sum(data))
                return
            self.increment_counter('MrJob Counters','reducer',1)
            yield (station, sum(counts))
        except Exception, e:
            yield (('error','reducer', str(e)), 1)
        
    def combiner(self, word, counts):
        try:
            self.increment_counter('MrJob Counters','combiner',1)
            yield (word, sum(counts))
        except Exception, e:
            yield (('error','combiner', str(e)), 1)
        
if __name__ == '__main__':
    count_TMIN_TMAX.run()
````

## 2) Partitioning: Efficient spatial data structure for geo data

We have around 9 million years of measurements. To fit a good model of the data, it would be smart to partition the data into smaller areas where the data might have similar patterns and fit a model on each area. If we blindly split the world into small tiles, we might end up having areas with few (or no) data points and areas with many data point, leading to overfitting in some parts of the world and underfitting in other parts. We would rater partition the data into areas of equal number of measurements.

We can do this by dividing the data points recursively using an R-tree ([Wikipedia](http://en.wikipedia.org/wiki/R-tree)).

### Constructing an R-tree

There are different heuristics defining how to split the data. Here we will just take a simple approach and split the data recursively into two parts with alternating latitude and longitude splits.

What this simple R-tree does is basicly:
```
    Split the data so that we have approximately equal number of measurements on each side (latitude or longitude)
    If both sides have more that 100 station-years of data then:
        Recurse, but change direction (form latitude to longitude or opposite)
```

````python
class rtree_node:
    def __init__(self, dataframe, split_direction='longitude', lable=0, level=0, threshold=365*100, original_data=merged):
        
        # Add labels
        original_data.ix[dataframe.index,'level_%d' % level] = lable
        
        # Split
        group_1, group_2, split_at = self.find_best_split(dataframe, split_direction)
        
        # Recursion base step
        if group_1.sum(axis=0)['measurement_count'] < threshold or group_2.sum(axis=0)['measurement_count'] < threshold:
            self.is_leaf = True
            self.data = dataframe
        # Recursive step
        else:
            self.lable = lable
            self.level = level
            self.is_leaf = False
            self.split_direction = split_direction
            self.split_at = split_at
            self.children = [rtree_node(group_1, self.switch_direction(split_direction), self.lable*2, self.level+1),
                             rtree_node(group_2, self.switch_direction(split_direction), self.lable*2+1, self.level+1)]
    
    # Helper functions

    def switch_direction(self, direction):
        latlon = ['latitude', 'longitude']
        return latlon[latlon.index(direction)-1]
        
    def find_best_split(self, dataframe, split):
        sorted_cumsum = dataframe.sort(split)['measurement_count'].cumsum()
        split_at = dataframe.sum(axis=0)['measurement_count']/2 # half of the sum
        group_1 = sorted_cumsum[sorted_cumsum <= split_at]
        group_2 = sorted_cumsum[sorted_cumsum > split_at]
        return dataframe.loc[group_1.index,:], dataframe.loc[group_2.index,:], split_at
    
    def print_(self, level=0):
        if self.is_leaf:
            print '    '*level, "# of measurements: ", sum([elem for elem in self.data['measurement_count']])
        else:
            self.children[0].print_(level+1)
            print '    '*level, 'split at:', self.split_at
            self.children[1].print_(level+1)
    
    def count(self):
        if self.is_leaf:
            return 1
        else:
            return self.children[0].count() + self.children[1].count()
````

### PCA

A PCA is basically just a Singular Value Decomposition (SVD) on the covariance matrix. The covariance matrix $cov$ is the sum of the outer products of each row $x_i$ in the dataframe:
$$UΣV^T = cov = \Sigma( x_i \cdot x_i^T )$$
Performing the SVD can be done with `Numpy`'s function `linalg.svd`.

We will do the calculations first on the small test dataset mentioned earlier:

````python
header=['station','measurement','year']+range(1,366)
local_weather = pd.read_csv(local_weather_data, header=None, names=header)
local_weather = local_weather.ix[local_weather['measurement'] == 'TMAX',]
local_weather = local_weather.dropna() # drop station-years where we do not have all measurements
local_weather.ix[:,1:365]=local_weather.ix[:,1:365]/10 #convert to celcius
local_weather.shape
lw = local_weather.ix[:,1:365]
lw.head()
````

Before calculating the covariance matrix, the data must be mean centered. This is why we subtract the mean before calculation the outer product.

````python
mean = lw.ix[:,1:365].mean()
````

````python
(rows, column) = lw.shape
cov = np.zeros([column, column])   # Covariance matrix = sum(row^T * row)
for i in range(rows):
    row = lw.iloc[i].ix[1:365] - mean # row i columns named 1 through 365
    outer_product = np.outer(row, row)
    cov += outer_product
cov
````

````python
U,D,V=np.linalg.svd(cov)
print "cov %s = U %s * D %s * V^(T) %s" % (shape(cov),shape(U),shape(D),shape(V))
````

The diagonal matrix $D$ is containing the eigenvalues. These can be interpreted as the ratio of the variance that is explained by the model.

````python
plot(cumsum(D[:])/sum(D))
grid()
````
