###Step 1: Compute station weight###
In the raw data, many stations do not have actual measurements, so we don't want to include these data for our analysis. There are many ways to define the valid measurements. For this analysis, we consider a year record with >50% non-empty measurements as valid.
As such, the station weight is computed by the number of its valid records. 

We use mapReduce to get the weight. The framework is similar with word frequency counter, just now the key is station and the value is its corresponding weight.


```python
%%writefile Stations_Weights.py
"""
collect the statistics for each station.
"""
#!/usr/bin/python
import re,pickle,base64,zlib
from sys import stderr
import sys

sys.path.append('/usr/lib/python2.6/dist-packages') # a hack because anaconda made mrjob unreachable
from mrjob.job import MRJob
from mrjob.protocol import *

import traceback
from functools import wraps
from sys import stderr



def loads(eVal):
    """ Decode a string into a value """
    return pickle.loads(zlib.decompress(base64.b64decode(eVal)))

def dumps(Value):
    """ Encode a value as a string """
    return base64.b64encode(zlib.compress(pickle.dumps(Value),9))

class MRWeather(MRJob):
    
    def check_integrity(self,meas,year,length,num_defined):
        year=int(year)
        if year<1000 or year > 2014: return False
        if meas=='': return False
        if length != 367: return False
        if num_defined < 188: return False   # have >50% non-empty measurements
        return True
    
    
    def mapper(self, _, line):
        try:
            self.increment_counter('Mrjob counter','mapper-all',1)
            elements=line.split(',')
            vector = elements[1:]
            meas=vector[0]   # measurement
            year=vector[1]   # year
            length=len(vector)  
            num_defined = sum([e!='' for e in vector[2:]])
            assert self.check_integrity(meas,year,length, num_defined)==True   # Data integrity checking
            yield(elements[0],1)    
        except Exception as e:
            self.increment_counter('Mrjob counter','mapper-error',1)
            pass
    
    def combiner(self, station, counts):
        self.increment_counter('MrJob Counters','combiner',1)
        yield (station, sum(counts))
     
    def reducer(self,station, counts):
        self.increment_counter('MrJob Counters','reducer',1)
        yield (station, sum(counts))
        
    
if __name__ == '__main__':
    MRWeather.run()
```

###Step 3: Partition the geographic data###
We apply the widely used K-D tree algorithm to the source data obtained in the last step. After partitioning, we should guaranttee each partition's total weights are similar. We tried different tree depth to see the variance of data points of different partitions. This piece of code generate 256 partitions, each of which has around 200~300 data points. There are a few partitions which contain far more data points than others because one stations in that area has been highly active, and we cannot do further partitioning inside one station.

After this, we are ready to start part 2, which applies PCA on all data points of each partition.

```python
#!/usr/bin/python
import sys,os
import pandas as pd
import numpy as np
import csv
from operator import itemgetter
import base64,pickle,zlib,sys
import pdb
from sys import stderr

def dump(key,Value,out=sys.stdout):
    out.write("%s\t%s\n" % (key,
base64.b64encode(zlib.compress(pickle.dumps(Value),9))))

class KDTree:
    def __init__(self):
        self.index = 0

    # Label the leaf of K-D tree
    def labelIndex(self, stations):
        n = 0
        count = 0
        for s in stations:
            #dump(s[0],{'longitude':s[1],'latitude':s[2], 'index': self.index})
            count += 1
        print "Index %d, data points %d"%(self.index, count)
        self.index += 1
    
    def getMedian(self, stations):
        total = 0
        count = 0
        for s in stations:
            total += s[3]
            count += 1
        weight = 0
        for i in range(0, count):
            weight += stations[i][3]
            if (weight >= total/2):
                return i

    def partition (self, stations, depth, n):
        if n == depth:
            self.labelIndex(stations)
            return

        axis = n % 2
        if (axis == 0):
            stations.sort(key=itemgetter(1)) # sort by longitude
        else:
            stations.sort(key=itemgetter(2))  # sort by latitude

        median = self.getMedian(stations)
        left = stations[0:median]
        right = stations[median+1:]
        self.partition(left, depth, n+1)
        self.partition(right, depth, n+1)

def main():
    tree = KDTree()
    f = open('/home/ubuntu/UCSD_BigData/data/weather/station_weight2.csv', 'rb') # opens the csv file
    reader = csv.reader(f)  # creates the reader object
    data = []
    next(reader)   # skip header
    for row in reader:
        try:
            station = row[0]
            longitude = float(row[1])
            latitude = float(row[2])
            weight =  float(row[3])
            data.append((station,longitude,latitude,weight))
        except Exception as e:
            #stderr.write('Error:')
            #stderr.write(str(e))
            pass
    tree.partition(data, depth = 8, n = 0)
    f.close()      # closing

if __name__=='__main__':
    main()
```
