We can just try the orignial PCA work on this small dataset, this program will find the $n$ which is the number of the vectors explains 99% variance

```python
%%writefile Test_Stats_PCA.py
#!/usr/bin/python
"""
collect the statistics for each station.
"""
import re,pickle,base64,zlib
from sys import stderr
import sys
import pandas as pd
import numpy as np
import sklearn as sk

sys.path.append('/usr/lib/python2.6/dist-packages') # a hack because anaconda made mrjob unreachable
from mrjob.job import MRJob
from mrjob.protocol import *

import traceback
from functools import wraps
from sys import stderr

class MRWeather(MRJob):

    def map_one(self,line):
        return line.split(',')
    
    def mapper(self, _, line):
        elements=self.map_one(line)
        #yield(elements[0],elements[1:])
        yield('sum',elements[1:])
            
    def check_integrity(self,meas,year,length):
        year=int(year)
        if year<1000 or year > 2014: return False
        if meas=='': return False
        if length != 367: return False
        return True
            
    def reducer(self, station, vectors):
        sumvec=np.zeros(365)
        matr=np.zeros((365,365))
        tt=0
        data={}
        for vector in vectors:
            meas=vector[0]
            year=vector[1]
            length=len(vector)
            number_defined=sum([e!='' for e in vector[2:]])
            if self.check_integrity(meas,year,length)==True and number_defined==365:
                if station=='sum':
                    strvec=vector[2:]
                    numvec=[int(ss) for ss in strvec]
                    tt=tt+1
                    data[tt]=numvec
        if station=='sum':
            for n in range(1,tt+1):
                sumvec=sumvec+np.array(data[n])
            sumlis=sumvec.tolist()
            meanlis=[float(n)/tt for n in sumlis]
            for n in range(1,tt+1):
                subtarr=np.array(data[n])-np.array(meanlis)
                matr+=np.outer(subtarr,subtarr)
            matr=np.divide(matr,tt)
            U,D,V=np.linalg.svd(matr)
            varexp=[]
            ptvar=0
            ttvar=sum(D)
            for n in range(0,365):
                ptvar+=D[n]
                if ptvar/ttvar>0.99:
                    break
            yield('PCA',n)
                              
if __name__ == '__main__':
    MRWeather.run()
```

Let's make partitions, this program is basically Shen Zhimo's idea, use recursive to find the partition.

```python
def takepartition(start,end,index,l):
    if end-start<=1:
        return l
    if index=='latitude':
        index2='longitude'
    else:
        index2='latitude'
    a=djoin[start:end].sort(index,inplace=False)
    djoin[start:end]=a
    temp=sum(djoin[start:end]['count'])
    #print temp
    if temp<100 :
        return l
    num=0
    for i in range(start,end):
        #print num
        num=num+djoin.ix[i,['count']].values
        if num>temp/2:
            if start==i:
                t1=(djoin.ix[i,[index]].values+djoin.ix[i+1,[index]].values)/2
                #print t1
                djoin.ix[start:i+1,[index+'2']]=t1
                l=takepartition(start,i+1,index2,l)
                djoin.ix[i+1:end,['area']]=l
                djoin.ix[i+1:end,[index+'1']]=t1
                l=l+1
                l=takepartition(i+1,end,index2,l)
                return l
            else:
                t1=(djoin.ix[i,[index]].values+djoin.ix[i-1,[index]].values)/2
                #print t1
                djoin.ix[start:i,[index+'2']]=t1
                l=takepartition(start,i,index2,l)
                djoin.ix[i:end,['area']]=l
                djoin.ix[i:end,[index+'1']]=t1
                l=l+1
                l=takepartition(i,end,index2,l)
                return l
```

Then let's do merge, the basic idea is:

If $X$ and $Y$ are two covariance matrices, $\alpha$ and $\beta$ are corresponding mean vectors, $n_1$ and $n_2$ are totally number of vectors contribute to $X$ and $Y$. Then the covariance matrix of these two data set together is:
$$
n_1 X+n_1+\alpha \alpha^T+n_2 Y+n_2+\beta \beta^T-\frac{1}{n_1+n_2} (\alpha+\beta)(\alpha+\beta)^T
$$

```python
%%writefile Test_Stats_Merge.py
#!/usr/bin/python
"""
collect the statistics for each station.
"""
import re,pickle,base64,zlib
from sys import stderr
import sys
import pandas as pd
import numpy as np
import sklearn as sk

sys.path.append('/usr/lib/python2.6/dist-packages') # a hack because anaconda made mrjob unreachable
from mrjob.job import MRJob
from mrjob.protocol import *

import traceback
from functools import wraps
from sys import stderr

class MRWeather(MRJob):
    
    def mapper(self, _, line):
        elements=line.split("\t")
        areaid=int(elements[0])
        tempstr=elements[1]
        tempstr=tempstr.replace("[","")
        tempstr=tempstr.replace("]","")
        tempstr=tempstr.replace(",","")
        templis=tempstr.split()
        templisnum=[float(ss) for ss in templis]
        for area2id in range(0,255):
            yield([areaid,area2id],templisnum)
            yield([area2id,areaid],templisnum)
            
    def adj(la11,la12,lo11,lo12,la21,la22,lo21,lo22):
        lis1=[[la11,lo11],[la11,lo12],[la12,lo11],[la12,lo12]]
        for ss in lis1:
            if ss[0]>=la21 and ss[1]>=lo21 and ss[0]<=la22 and ss[1]<=lo22:
                return True
        return False

    def reducer(self, area, vectors):
        id1=area[0]
        id2=area[1]
        tt=0
        data={}
        for vec in vectors:
            data[tt]=vec
            tt=tt+1
        vec1=data[0]
        vec2=data[1]
        if id1<id2:
            nn1=vec1[0]
            la11=vec1[1]
            la12=vec1[2]
            lo11=vec1[3]
            lo12=vec1[4]
            nn2=vec1[0]
            la21=vec1[1]
            la22=vec1[2]
            lo21=vec1[3]
            lo22=vec1[4]
            if self.adj(la11,la12,lo11,lo12,la21,la22,lo21,lo22):
                mealis1=vec1[5:370]
                covlis1=vec1[371:]
                mealis2=vec2[5:370]
                covlis2=vec2[371:]
                meaarr1=np.array(mealis1)
                meaarr2=np.array(mealis2)
                covarr1=np.array(covlis1)
                covarr2=np.array(covlis2)
                covmat1=covarr1.reshape((365,365))
                covmat2=covarr2.reshape((365,365))
                temp1=covmat1+np.outer(meaarr1,meaarr1)
                temp2=covmat2+np.outer(meaarr2,meaarr2)
                newmea=np.divide(meaarr1+meaarr2,nn1+nn2)
                matr=np.multiply(temp1,nn1)+np.multiply(temp2,nn2)-np.multiply(newmea,nn1+nn2)
                matr=np.divide(matr,nn1+nn2)
                U,D,V=np.linalg.svd(matr)
                varexp=[]
                ptvar=0
                ttvar=sum(D)
                for n in range(0,365):
                    ptvar+=D[n]
                    if ptvar/ttvar>0.99:
                        break
                yield(area,n)
                              
if __name__ == '__main__':
    MRWeather.run()
```
