## 1. Measurement and weather station statistics

The five core element types are:

       PRCP = Precipitation (tenths of mm)
       SNOW = Snowfall (mm)
	   SNWD = Snow depth (mm)
       TMAX = Maximum temperature (tenths of degrees C)
       TMIN = Minimum temperature (tenths of degrees C)

In the following, we will focus on the elements ``TMIN`` and ``TMAX``, as they should have the most records across all weather stations. A larger and well distributed sample size allows us to get more meaningful statistics about the data. Moreover, it will be desirable when partitioning the data and running the principal component analysis (PCA).

In order to analyze the statistics of the data, I implemented two jobs. The first one is ``mr_weather_count_measurements.py``. This job counts the number of measurements of a specified element for either every station or year.

```python
%%writefile mr_weather_count_measurements.py
"""
count the number of measurements of each type
"""
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr

elementType = 'TMAX'
key = 2   # 0 for station, 2 for year

class MRWeatherCountMeasurements(MRJob):

    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper',1)
            elements=line.split(',')
            measurements_per_line = 0
            if elements[1]==elementType:    
                for j in range(365):
                    if isinstance(elements[j+3], basestring):
                        measurements_per_line += 1        
            if elements[0]!='station':
                out=(elements[key],measurements_per_line)
            else:
                out=('header',1)
        except Exception, e:
            stderr.write('Error in line:\n'+line)
            stderr.write(e)
            self.increment_counter('MrJob Counters','mapper-error',1)
            out=('error',1)
        finally:
            yield out
            
    def combiner(self, word, counts):
        self.increment_counter('MrJob Counters','combiner',1)
        yield (word, sum(counts))

    def reducer(self, word, counts):
        self.increment_counter('MrJob Counters','reducer',1)
        yield (word, sum(counts))

if __name__ == '__main__':
    MRWeatherCountMeasurements.run()
```

The second job is ``mr_weather_count_365.py``. Similar to the previous job, it counts the number of measurements of a specified element for either every station or year. However, it only considers measurement series, where values for all 365 days are available.

```python
%%writefile mr_weather_count_365.py
"""
count the number of measurement series with non-NaN values for all 365 days
"""
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr
import numpy as np

elementType = 'TMAX'
key = 0   # 0 for station, 2 for year

class MRWeatherCountMeasurements(MRJob):

    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper',1)
            elements=line.split(',')
            mvalues = np.array(elements[3:])
            if elements[0]!='station':
                if elements[1]==elementType:    
                    if np.where(mvalues=='',True,False).any():
                        out=(elements[key],0)
                    else:
                        out=(elements[key],365)
                else:
                    out=(elements[key],0)
            else:
                out=('header',1)
        except Exception, e:
            stderr.write('Error in line:\n'+line)
            stderr.write(e)
            self.increment_counter('MrJob Counters','mapper-error',1)
            out=('error',1)
        finally:
            yield out
            
    def combiner(self, word, counts):
        self.increment_counter('MrJob Counters','combiner',1)
        yield (word, sum(counts))

    def reducer(self, word, counts):
        self.increment_counter('MrJob Counters','reducer',1)
        yield (word, sum(counts))

if __name__ == '__main__':
    MRWeatherCountMeasurements.run()
```

#### Select stations for given boundary ####

```python
from mpl_toolkits.basemap import Basemap, cm
import matplotlib.pyplot as plt

def select_stations(dataframe,bounds=[-80,80,-180,180]):
    lons=dataframe.ix[:,'longitude'].values
    lats=dataframe.ix[:,'latitude'].values 
    select=(lons>bounds[2]) * (lons<bounds[3])*(lats>bounds[0])*(lats<bounds[1])
    print 'Total stations selected: ', sum(select)
    return(lons[select],lats[select],select)
```

```python
complete_bounds = [-80,80,-180,180]
(latmin, latmax, lonmin, lonmax) = complete_bounds
(lons,lats,select)=select_stations(dfComplete,complete_bounds)
dfComplete = dfComplete[select]
```

#### Density of weather stations ####

```python
plt.figure(figsize=(15,10),dpi=300)
m = Basemap(projection='merc',llcrnrlat=latmin,urcrnrlat=latmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
m.drawcoastlines()
# draw parallels and meridians.
parallels = np.arange(-80,81,10.)
m.drawparallels(parallels,labels=[False,True,True,False])
meridians = np.arange(10.,351.,20.)
m.drawmeridians(meridians,labels=[True,False,False,True])

# compute density of weather stations via 2D histogram
x, y = m(lons,lats)
xyrange = [[min(x),max(x)],[min(y),max(y)]] # data range
nx = 500 # number of bins along x
ny = 500 # number of bins along y
EarthSurface = 510e6 # surface of earth in km^2
binArea = EarthSurface/(nx*ny)
# generate histogram
hh, xedges, yedges = np.histogram2d(x, y, range=xyrange, bins=[nx,ny])
hh[hh < 1] = np.nan # fill the areas with no stations by NaNs

m.imshow(hh.T,extent=np.array(xyrange).flatten(),interpolation='nearest', cmap='jet', vmin=1, vmax=16)

plt.colorbar()
plt.title('No. of stations per bin area ('+str(round(binArea))+'km^2)')
plt.show()
```

#### Density of measurements ####

```python
def plot_measurements(names,xbins=250,ybins=250,vmin=0,vmax=20000,log=False):
    Npar = len(names)
    data = np.zeros((Npar,len(lons)))
    for i in range(Npar):
        data[i,:] = np.reshape(dfComplete.ix[:,names[i]].values[select],[1,len(lons)])

    fig=plt.figure(figsize=(16,10),dpi=300)
    for i in range(Npar):
        subplot(2,2,i)
        #draw world map
        m = Basemap(projection='merc',llcrnrlat=latmin,urcrnrlat=latmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
        m.drawcoastlines() 
        parallels = np.arange(latmin,latmax,10) ###
        m.drawparallels(parallels,labels=[False,True,True,False])
        meridians = np.arange(lonmin,lonmax,20) ###
        m.drawmeridians(meridians,labels=[True,False,False,True])
        x, y = m(lons,lats)

        xyrange = [[min(x),max(x)],[min(y),max(y)]] # data range
        nx = xbins # number of bins along x
        ny = ybins # number of bins along y
        # generate histogram
        hh, xedges, yedges = np.histogram2d(x, y, range=xyrange, bins=[nx,ny],weights=data[i,:])
        hh[hh < 1] = np.nan # fill the areas with no stations by NaNs
        if log==True:
            hh = np.log10(hh)
        #m.imshow(hh.T,extent=np.array(xyrange).flatten(),interpolation='nearest', cmap='jet', vmin=vmin, vmax=vmax)
        #plot data points
        plt.scatter(x, y, c=data[i,:], s=2, linewidth=0, vmin=vmin, vmax=vmax, cmap=plt.cm.jet)
        plt.colorbar()
        plt.title(names[i])
    plt.tight_layout()
```

Plotting the measurements density in a log scale, we find that most of the measurement data is concentrated in North America. The density in the other continents is at least one order of magnitude lower.

```python
names = ['N_TMIN','N_TMAX','N_TMIN365','N_TMAX365']
plot_measurements(names,250,250,1,6,log=True)
```
