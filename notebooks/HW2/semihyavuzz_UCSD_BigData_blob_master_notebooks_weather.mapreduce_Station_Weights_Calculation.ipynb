```python
%%writefile mr_weight.py
from mrjob.job import MRJob

class MRStationWeightsCount(MRJob):
  def mapper(self, _, line):
    arr = line.strip().split(',')
    st = arr[0]
    measurement_type = arr[1]
    measurement_year = arr[2]

    measurement_data, record_number = [], 0
    itr=0
    number_days=365
    threshold=0.2
    while itr<number_days:
        t = arr[3 + itr]
        try:
            x=int(t)
        except ValueError:
            x=None
        
        if x is not None: record_number += 1
        measurement_data.append(x)
        itr+=1
    
    if float(record_number) / number_days < threshold:
            output=None
    else:
            output= (st, measurement_year, measurement_type, measurement_data, record_number)
    
    if output is not None:
      station, year, measurement, data, records = output
      if measurement == 'TMIN' or measurement == 'TMAX':
        yield station, 1

  def combiner(self, key, vals):
    yield key, sum(vals)

  def reducer(self, key, vals):
    yield key, sum(vals)

if __name__ == '__main__':
  MRStationWeightsCount.run()
```
