### 1. Statistics of NaN Distribution of TMIN/TMAX Data ###

Since we only care about the TMIN and TMAX data, first we use MapReduce to get the NaN value distribution of the TMAX and TMIN data. The following script calculates the number of non-NaN values of each TMIN/TMAX row.

```python
%%writefile Stations_TMin_TMax_Statistics.py
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr

class MRWeather(MRJob):
    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper-all',1)
            elements=line.split(',')
            if elements[0]=='station':
                yield (('header',1), 1)
            else:
                if elements[1] == 'TMIN':
                    num_defined = sum([e!='' for e in elements[3:]])
                    yield (('TMIN', num_defined), 1)
                elif elements[1] == 'TMAX':
                    num_defined = sum([e!='' for e in elements[3:]])
                    yield (('TMAX', num_defined), 1)
        except Exception, e:
            stderr.write('Error in line:\n'+line)
            stderr.write(e)
            self.increment_counter('MrJob Counters','mapper-error',1)
            yield (('error', 1),  1)

    def combiner(self, key, values):
        yield (key, sum(values))
    
    def reducer(self, key, values):
        self.increment_counter('MrJob Counters','reducer',1)
        yield (key, sum(values))
        
if __name__ == '__main__':
    MRWeather.run()
```

### 2. Statistics of Station-Year which has both TMAX and TMIN

The following code calculates how much data there are for stations which have both TMAX and TMIN data for a certain year. 

```python
%%writefile Stations_TMin_TMax_BOTH_Statistics.py
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr

threshold = 350

class MRWeather(MRJob):
    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper-all',1)
            elements=line.split(',')
            num_defined = sum([e!='' for e in elements[3:]])
            if num_defined >= threshold:
                if elements[1] == 'TMIN':
                    yield ((elements[0],elements[2]), 1)
                elif elements[1] == 'TMAX':
                    yield ((elements[0],elements[2]), 1)
                    
        except Exception, e:
            stderr.write('Error in line:\n'+line)
            stderr.write(e)
            self.increment_counter('MrJob Counters','mapper-error',1)

    def reducer(self, key, values):
        if sum(values) == 2: # have both TMAX and TMIN
            yield ("valid", 1)
    
    def final_reducer(self, key, values):
        self.increment_counter('MrJob Counters','reducer',1)
        yield (key, sum(values))
    
    def steps(self):
        return [self.mr(mapper=self.mapper, reducer=self.reducer),
                 self.mr(reducer=self.final_reducer)]
    
if __name__ == '__main__':
    MRWeather.run()
```

### 3. Filter and Clean Data

The following code uses MapReduce to filter and clean the weather data. The TMAX and TMIN data of stations for a year are kept if the number of non-empty values is more than 350. All other data rows are dropped. Additionally, TMAX and TMIN data for each station-year is concatenated as one row. Empty values are replaced with 0. The final data is stored locally and then uploaded to my S3 dir for further analysis.

```python
%%writefile Weather_Filter.py
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
from numpy import *
import re
from sys import stderr

threshold = 350

class MRWeather(MRJob):
    
    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper-all',1)
            elements=line.split(',')
            num_defined = sum([e!='' for e in elements[3:]])
            if num_defined >= threshold:
                if elements[1] == 'TMIN':
                    yield (elements[0],elements[2]), ('TMIN', elements[3:])
                elif elements[1] == 'TMAX':
                    yield (elements[0],elements[2]), ('TMAX', elements[3:])
                    
        except Exception, e:
            stderr.write('Error in line:\n'+line)
            stderr.write(e)
            self.increment_counter('MrJob Counters','mapper-error',1)

    def reducer(self, key, values):
        self.increment_counter('MrJob Counters','reducer-all',1)
        lst = list(values)
        if len(lst) == 2:
            if lst[0][0]=='TMIN':
                data = lst[0][1]+lst[1][1]
            else:
                data = lst[1][1]+lst[0][1]
            good_data = []
            for e in data:
                if e != '':
                    good_data.append(int64(e))
                else:
                    good_data.append(0)
            yield (key, good_data)
    
if __name__ == '__main__':
    MRWeather.run()
```

### 4. Calculate the weight of each station 

The weight of each station is defined as the total number of non-empty measurements in my filtered data, which is s3://ruininghe/Weather/weather.csv.

```python
%%writefile Stations_Weight.py
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr

class MRWeather(MRJob):    
    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper-all',1)
            station = line.split('[')[1].split(',')[0]
            station = station.replace('"', '')
            data = line.split('[')[2].split(']')[0].split(',')
            num_defined = 0
            for e in data:
                if int(e) != 0:
                    num_defined = num_defined + 1
            yield (station, num_defined)
                    
        except Exception, e:
            stderr.write('Error in line:\n'+line)
            stderr.write(e)
            self.increment_counter('MrJob Counters','mapper-error',1)

    def combiner(self, key, values):
            yield key, sum(values)
            
    def reducer(self, key, values):
            yield key, sum(values)
    
if __name__ == '__main__':
    MRWeather.run()
```

### 6. Partition Stations With KDTree

The following code partitions stations with KDTree. Each partition is performed either by latitude or longitude. The parameter levels means how many levels the final KD tree has. I take levels as 9 because there will be 512 initial groups and each groups will have about sevel thousand rows, which is reasonable for the further PCA analysis.

```python
import pandas as pd

class KDTree(object):

    def __init__(self, df, levels):
        self.levels = levels
        self.df = df

    def get_id(self, bits):
        return sum(bit * (2**i) for bit, i in zip(bits, range(len(bits))))

    def get_partition(self, partition):
        if partition == 'latitude':
            return 'longitude'
        return 'latitude'

    def execute(self, df, level, partition='latitude', N=90, S=-90, W=-180, E=180, bits=[]):
        if level <= 0:
            id = self.get_id(bits)
            df['id'] = id
            region = {'id': id,
                      'N': N, 'S': S, 'W': W, 'E': E,
                      'weight': df.weight.sum()}
            return df, region

        df = df.sort(partition)
        df['cumulative_weight'] = df.weight.cumsum()
        threshold = df.weight.sum()/2
        tree_id = 'tree' + str(level)
        df[tree_id] = df.cumulative_weight > threshold
        del df['cumulative_weight']
        grouped_df = df.groupby(tree_id)

        group_below_threshold = pd.DataFrame()
        group_above_threshold = pd.DataFrame()
        for flag, group in grouped_df:
            if flag: group_above_threshold = group
            else: group_below_threshold = group

        min_Value = group_above_threshold[partition].min()
        max_Value = group_below_threshold[partition].max()

        border = (max_Value + min_Value)/2

        children_df = []
        children_leaves = []

        for flag, child in grouped_df:
            if partition == 'latitude':
                if flag:
                    S = border
                else:
                    N = border
            else:
                if flag:
                    W = border
                else:
                    E = border
            child_df, child_leaves = self.execute(child,
                                                  level-1,
                                                  self.get_partition(partition),
                                                  N, S, W, E,
                                                  [flag]+bits)
            children_df.append(child_df)
            children_leaves.append(child_leaves)

        return pd.concat(children_df), children_leaves
```
