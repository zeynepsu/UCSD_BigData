```python
%%writefile weather_00.py
#!/usr/bin/python
"""
count the number of measurement for each year
"""
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr

class MRWeather(MRJob):

    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper-all',1)
            elements=line.split(',')
            if elements[0]=='station':
                out=('0000',(1,1))
            else:
                if elements.count('') > 75: 
                    # At least 20% of cells are filled in
                    out=(elements[2],(1,1))
                else:
                    out=(elements[2],(1,0))
        except Exception, e:
            stderr.write('Error in line:\n'+line)
            stderr.write(e)
            self.increment_counter('MrJob Counters','mapper-error',1)
            out=('error',(1,1))

        finally:
            yield out
            
    def combiner(self, word, counts):
        self.increment_counter('MrJob Counters','combiner',1)
        (a,b) = zip(*counts)
        yield (word, (sum(a),sum(b)))
        
        
        #logfile.write('combiner '+word+' ['+','.join([str(c) for c in l_counts])+']='+str(S)+'\n')
        #yield (word, S)

    def reducer(self, word, counts):
        self.increment_counter('MrJob Counters','reducer',1)
        
        (a,b) = zip(*counts)
        yield (word, (sum(a),sum(b)))
        
        #l_counts=[c for c in counts]  # extract list from iterator
        #S=sum(l_counts)
        #logfile.write('reducer '+word+' ['+','.join([str(c) for c in l_counts])+']='+str(S)+'\n')
        #yield (word, S)

if __name__ == '__main__':
    MRWeather.run()
```

###Common printer

```python
# Plot a file, where each line is of the form:
# <year> <number of measurements>

def plot_sp_dist(name):
    from matplotlib import pyplot as plt
    import numpy as np
    with open(name) as f:
        #yrs = np.loadtxt(f, dtype = int, delimiter="\t", converters = {0: lambda s: int(s.replace('"','')) })
        regexp = r'\s*\"(\d+)\"\s+\[(\d+),\s*(\d+)\]\s*'  # match [digits, whitespace, anything]
        yrs = np.fromregex(f, regexp, [('num1', np.int64),('num2', np.int64),('num3', np.int64)])

    # Create an array from the parsed lines
    arr = np.array([ np.array([a, b, c]) for (a,b,c) in np.array(yrs) ])
    
    # Sort the array by year
    arr = arr[arr[:,0].argsort()]

    print 'Total number of measurements:', sum(arr[:,1])
    print
    print 'For the figures below:'
    print '  * red line: number of entries with at least one measurement per year'
    print '  * green line: number of entries with measurements for at least 20% of the days of the year'
    print
    
    # Filter years later than 1700
    arr1700 = arr[(1700 < arr[:,0]) & (arr[:,0] < 1860)]
    
    # Filter years later than 1900
    arr1900 = arr[1860 < arr[:,0]]

    plt.close('all')
    fig = plt.figure(figsize=[12,5])
    
    ax1 = plt.subplot(121)
    p1  = ax1.plot(arr1700[:,0], arr1700[:,1], 'r', arr1700[:,0], arr1700[:,2], 'g')
    #legend([p1, p1], ["line 2", "line 1"])

    ax1.set_xlabel("Year", fontsize=13)
    ax1.set_ylabel("#measurements", fontsize=13)
    ax1.set_title("Years: 1750-1860", fontsize=13)
    ax1.grid()
    
    ax2 = plt.subplot(122)
    ax2.plot(arr1900[:,0], arr1900[:,1],'r', arr1900[:,0], arr1900[:,2],'g')
    ax2.set_xlabel("Year", fontsize=13)
    ax2.set_ylabel("#measurements", fontsize=13)
    ax2.set_title("Years: 1860-2014", fontsize=13)
    
    ax2.grid()

    plt.tight_layout()
```
