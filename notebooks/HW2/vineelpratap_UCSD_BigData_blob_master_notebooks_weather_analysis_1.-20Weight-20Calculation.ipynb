### Calculating "counts" for the measurements

<font color='blue'> For every station, the following map-reduce job  will give the number of readings defined out of 365 possible days for each (measurement, year) pair whenever available. </font>

```python
%%writefile Stations_Statistics.py
#!/usr/bin/python
"""
collect the statistics for each station.
"""
import re,pickle,base64,zlib
from sys import stderr
import sys

sys.path.append('/usr/lib/python2.6/dist-packages') # a hack because anaconda made mrjob unreachable
from mrjob.job import MRJob
from mrjob.protocol import *

import traceback
from functools import wraps
from sys import stderr

"""this decorator is intended for decorating a function, not a
generator.  Therefore to use it in the context of mrjob, the generator
should call a function that handles a single input records, and that
function should be decorated.

The reason is that if a generator throws an exception it exits and
cannot process any more records.

"""
def ECatch(func):
    f_name=func.__name__
    #@wraps(func) cd
    def inner(self,*args,**kwargs):
        try:
            self.increment_counter(self.__class__.__name__,'total in '+f_name,1)
            return func(self,*args,**kwargs)
        except Exception as e:
            self.increment_counter(self.__class__.__name__,'errors in '+f_name,1)
            stderr.write('Error:')
            stderr.write(str(e))
            traceback.print_exc(file=stderr)
            stderr.write('Arguments were %s, %s\n'%(args,kwargs))
            pass
    return inner        

"""
Functions for encoding and decoding arbitrary object into ascii 
so that they can be passed through the hadoop streaming interface.
"""
def loads(eVal):
    """ Decode a string into a value """
    return pickle.loads(zlib.decompress(base64.b64decode(eVal)))

def dumps(Value):
    """ Encode a value as a string """
    return base64.b64encode(zlib.compress(pickle.dumps(Value),9))

meas_required = ['TMIN', 'TMAX', 'PRCP', 'SNWD', 'SNOW'] #core readings
min_year = 1900 #don't consider readings before this year


class MRWeather(MRJob):
    INTERNAL_PROTOCOL = mrjob.protocol.JSONProtocol
    OUTPUT_PROTOCOL = mrjob.protocol.JSONProtocol
    @ECatch
    def map_one(self,line):
        return line.split(',')
    
    def mapper(self, _, line):
        elements=self.map_one(line)
        yield(elements[0],elements[1:])
            
    def check_integrity(self,meas,year,length):
        if year<1000 or year > 2014: return False
        if meas=='': return False
        if length != 367: return False
        return True
    
    @ECatch
    def reduce_one(self,S,vector):
        meas=vector[0]
        year=int(vector[1])
        length=len(vector)
        number_defined=sum([e!='' for e in vector[2:]])
        assert self.check_integrity(meas,year,length)==True
        if (meas in meas_required and year>= min_year):
            S[(meas,year)]=number_defined
        
    def reducer(self, station, vectors):
        S={}
        for vector in vectors:
            self.reduce_one(S,vector)
        yield(station,dumps(S))
                              
if __name__ == '__main__':
    MRWeather.run()
```

<font color='blue'> Loading the output of mrjob into a dictionary </font>

```python
def loads(eVal):
    """ Decode a string into a value """
    return pickle.loads(zlib.decompress(base64.b64decode(eVal)))

def load_statsdata_to_dict(loc):
    stn_data, l = {}, 0
    f = open(loc)
    for line in f:
        l = l+1 
        stn, data_enc = line.split()
        stn_data[stn[1:-1]] = loads(data_enc[1:-1])
    
        if(l%(num_stations/10) == 0): print str((l*10)/num_stations+1)+"0% done"
        
    return stn_data

stats = load_statsdata_to_dict('StationStatistics')
```

<font color='blue'> Now we define reading in an year to be valid if the readings are available for more than 50% days of the year. <br />
The following code computes the number of valid years for each measurement. <br /> </font>

```python
meas_required = ['TMIN', 'TMAX', 'PRCP', 'SNWD', 'SNOW'] #core readings
#appending columns for counts
for meas in meas_required:
    Data_Stations[meas+'_count'] = 0
    
Data_Stations['maxmin_count'] = 0 
```

```python
valid_years= dict([(stn,[]) for stn in stations]) #store valid years for which both TMIN and TMAX are available

for stn in stations:
    cur_stn_data = stats[stn] if stats.has_key(stn) else {}
    cnt = {}
    for meas in meas_required : 
        cnt[meas] = 0
        
    year_set = set([]) 
    
    for (meas,yr) in cur_stn_data.keys():
        if(cur_stn_data[(meas,yr)] > 180):
            cnt[meas] = cnt[meas]+1
        if(meas=='TMIN' or meas=='TMAX'): year_set.add(yr)
    
    #this computes the number of valid years for which both TMIN and TMAX readings are available for the station
    maxmin_cnt = 0
    for yr in year_set:
        mx_cnt = cur_stn_data[('TMAX',yr)] if cur_stn_data.has_key(('TMAX',yr)) else 0
        mn_cnt = cur_stn_data[('TMIN',yr)] if cur_stn_data.has_key(('TMIN',yr)) else 0
        if(mx_cnt>=180 and mn_cnt>=180): 
            maxmin_cnt = maxmin_cnt + 1 
            valid_years[stn].append(yr) 
            
    for meas in meas_required: 
        Data_Stations[meas+'_count'][stn] = cnt[meas]  
    
    Data_Stations['maxmin_count'][stn] = maxmin_cnt
```
