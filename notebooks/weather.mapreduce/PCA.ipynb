```python
import urllib
opener = urllib.URLopener()
url = "https://s3.amazonaws.com/weather.raw_data/gsod.all.tsv_8"
#req = urllib2.Request(url, headers ={'User-Agent':'Chrome'})
#raw = urllib2.urlopen(req).read()
myfile = opener.open(url)
myfile
```

```python
#!/bin/env python
import os,sys,re,pickle,coding
from numpy import *


#lines = [line.strip() for line in open(home_dir+'/data/gsod.all.tsv_1024')]
#lines = [line.strip() for line in open('s3://zsubucket/gsod.all.tsv_1024')]
lines = [line.strip() for line in myfile]
oldyear = lines[1].strip().split(',')[2][:4]
oldid = lines[1].strip().split(',')[0]
meantemp = {}
mintemp = {}
maxtemp = {}
absmintemp = {}
absmaxtemp = {}
count = 0
summean = 0
summin = 0
summax = 0
absmin = 9999
absmax = -9999
for line in lines:
    line = re.sub('[!@#$*]', '', line)
    arr = line.strip().split(',')
    id = arr[0]
    date = arr[2]
    year = date[:4]
    month = date[4:6]
    day = date[6:8]
    meantempday = arr[3]
    maxtempday = arr[17]
    mintempday = arr[18]
    
    if oldyear==year and id==oldid:
        if float(meantempday)!=9999.9 and float(mintempday)!=9999.9 and float(maxtempday)!=9999.9:
            summean+=float(meantempday)
            summin+=float(mintempday)
            summax+=float(maxtempday)
            count=count+1
            if float(mintempday)<absmin:
                absmin=float(mintempday)
            if float(maxtempday)>absmax:
                absmin=float(mintempday)
    else:
        if count==0:
            count=1
        meantemp[id+year]=summean/count
        mintemp[id+year]=summin/count
        maxtemp[id+year]=summax/count
        absmintemp[id+year]=absmin
        absmaxtemp[id+year]=absmax
        count = 0
        summean = 0
        summin = 0
        summax = 0
        absmin = 9999
        absmax = -9999
        
    oldyear = year
    oldid=id
    
for key,val in mintemp.items():
    print key, "=>", val
    
eat = pd.DataFrame(meantemp.items(), columns=['idyear', 'temp'])
output = open('meantemp.pkl', 'wb')
pickle.dump(eat, output)
output.close()
eat = pd.DataFrame(mintemp.items(), columns=['idyear', 'temp'])
output = open('mintemp.pkl', 'wb')
pickle.dump(eat, output)
output.close()
eat = pd.DataFrame(maxtemp.items(), columns=['idyear', 'temp'])
output = open('maxtemp.pkl', 'wb')
pickle.dump(eat, output)
output.close()
eat = pd.DataFrame(absmintemp.items(), columns=['idyear', 'temp'])
output = open('absmintemp.pkl', 'wb')
pickle.dump(eat, output)
output.close()
eat = pd.DataFrame(absmaxtemp.items(), columns=['idyear', 'temp'])
output = open('absmaxtemp.pkl', 'wb')
pickle.dump(eat, output)
output.close()
```

```python
#!/bin/env python
import os,sys,re,pickle,coding
from numpy import *
import pandas as pd

lines = [line.strip() for line in locfile]
count=0

res = pd.DataFrame(columns=('stid', 'stlat', 'stlon'))
for line in lines:
    count+=1
    if count<21 :
        continue
    arr = line.strip().split(',')
    id = arr[0]
    lat = arr[7]
    lon = arr[8]

    row = pd.DataFrame([dict(stid=id, stlat=lat, stlon=lon), ])
    res = res.append(row, ignore_index=True)
res
output = open('idlatlon.pkl', 'wb')
pickle.dump(res, output)
output.close()
```
