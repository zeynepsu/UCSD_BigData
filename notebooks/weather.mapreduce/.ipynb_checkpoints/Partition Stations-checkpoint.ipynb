In this notebook, I first partition the stations using the Delaunay tessellation of Scipy and then taking into account the weights, I use the vertices that the Delaunay gives as indices to the original stations table that I constructed and I use all of these to construct the final regions and visualize them.

```python
#!/bin/env python
import os,sys,re,pickle,coding
from numpy import *
```

```python
stations=pickle.load(open('stationswithweights', 'rb'))
stations
```

```python
stations.sort(['latitude', 'longitude'], ascending=[True, True], inplace=True)
stations
```

```python
from scipy.spatial import Delaunay
from matplotlib.pyplot import plot

stationspartition = stations[['latitude','longitude']]
stationsdel = Delaunay(stationspartition)
triplot(stations['longitude'],stations['latitude'],stationsdel.simplices)
```

```python
stationsdel.vertices
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

lonmin=-180;lonmax=180;latsmin=-80;latsmax=80;
plt.figure(figsize=(15,10),dpi=300)
m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
m.drawcoastlines(linewidth=0.5)#,color='grey')

parallels = np.arange(-80,81,10.)
m.drawparallels(parallels,labels=[False,True,True,False])
meridians = np.arange(10.,351.,20.)
m.drawmeridians(meridians,labels=[True,False,False,True])

lats= stations.loc[:,'latitude'].values
lons= stations.loc[:,'longitude'].values
ll=len(lons)
selection = stations.loc[:,'area'].values<256
lats = lats[selection]
lons = lons[selection]
x, y = m(lons,lats)
z= stations.loc[:,'area'].values
z = z[selection]
m.scatter(x, y, c=z, s=20, cmap=plt.cm.jet, edgecolors='None')
cbar = plt.colorbar()
#plt.clim(-16,32)
cbar.ax.set_ylabel('area index')
plt.title('Eligible weather stations')
plt.show()
```
